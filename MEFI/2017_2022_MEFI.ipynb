````python
import pandas as pd
from rapidfuzz import process, fuzz

# ---------- helpers ----------
def z5(x):
    if pd.isna(x): return pd.NA
    s = str(x).strip()
    if s == "" or s.lower() == "nan": return pd.NA
    # strip decimals like '17410.0'
    if s.replace(".","",1).isdigit():
        s = str(int(float(s)))
    return s.zfill(5)

def norm(s):
    if pd.isna(s): return ""
    s = str(s).lower()
    for ch in ["–","—","(",")",".",","]:
        s = s.replace(ch, " ")
    s = s.replace("-", " ")
    return " ".join(s.split())

# ---------- load SMART 2017/2022 ----------
s17 = pd.read_csv("/content/drive/MyDrive/Saksham_MEFI/SMART_2017_individual.csv", dtype=str, low_memory=False)
s22 = pd.read_csv("/content/drive/MyDrive/Saksham_MEFI/SMART_2022_individual.csv", dtype=str, low_memory=False)
for df in (s17, s22):
    if "CBSA_CODE" in df.columns:
        df["CBSA_CODE"] = df["CBSA_CODE"].apply(z5)
    if "MMSANAME" in df.columns:
        df["MMSANAME_N"] = df["MMSANAME"].apply(norm)

smart_cbsa = pd.Series(pd.concat([s17.get("CBSA_CODE"), s22.get("CBSA_CODE")], ignore_index=True), name="CBSA_CODE").dropna().unique().tolist()
smart_cbsa = sorted(set(smart_cbsa))

# ---------- load MEFI 2017 roster ----------
mefi = pd.read_excel("/content/drive/MyDrive/Saksham_MEFI/MEFI+2022+dataset+for+researchers,+1-5-23 (4).xlsx",
                     sheet_name="scores (panel)")
mefi["CBSA Code"] = mefi["CBSA Code"].apply(z5)
mefi2017 = mefi[mefi["year"]==2017].copy()
mefi2017_codes = set(mefi2017["CBSA Code"].dropna())

# ---------- load OMB List 1 (older + 2023) ----------
l_old = pd.read_excel("/content/drive/MyDrive/Saksham_MEFI/list1.xls", sheet_name="List 1", header=0)
l_new = pd.read_excel("/content/drive/MyDrive/Saksham_MEFI/list1_2023.xlsx", sheet_name="List 1", header=0)

for df in (l_old, l_new):
    df.rename(columns={
        "CBSA Code": "CBSA_CODE",
        "CBSA Title": "CBSA_TITLE",
        "Metropolitan/Micropolitan Statistical Area": "MM_TYPE",
        "State Name": "STATE_NAME"
    }, inplace=True, errors="ignore")
    df["CBSA_CODE"] = df["CBSA_CODE"].apply(z5)
    df["CBSA_TITLE_N"] = df["CBSA_TITLE"].apply(norm)

old_codes = set(l_old["CBSA_CODE"].dropna())
new_codes = set(l_new["CBSA_CODE"].dropna())

# ---------- step 1: SMART codes not in MEFI 2017 ----------
smart_unknown = sorted(set(smart_cbsa) - mefi2017_codes)
unknown_df = pd.DataFrame({"legacy_cbsa_code": smart_unknown})

# attach titles/types where available
unknown_df = unknown_df.merge(l_old[["CBSA_CODE","CBSA_TITLE","CBSA_TITLE_N","MM_TYPE","STATE_NAME"]],
                              left_on="legacy_cbsa_code", right_on="CBSA_CODE", how="left") \
                       .drop(columns=["CBSA_CODE"]) \
                       .rename(columns={"CBSA_TITLE":"legacy_title","CBSA_TITLE_N":"legacy_title_n","MM_TYPE":"legacy_mm_type","STATE_NAME":"legacy_state"})

unknown_df = unknown_df.merge(l_new[["CBSA_CODE","CBSA_TITLE","CBSA_TITLE_N","MM_TYPE","STATE_NAME"]],
                              left_on="legacy_cbsa_code", right_on="CBSA_CODE", how="left", suffixes=("","_2023")) \
                       .drop(columns=["CBSA_CODE"]) \
                       .rename(columns={"CBSA_TITLE":"new_title_hit","CBSA_TITLE_N":"new_title_hit_n","MM_TYPE":"new_mm_type","STATE_NAME":"new_state"})

# ---------- step 2: classify ----------
def classify(row):
    code = row["legacy_cbsa_code"]
    in_old = code in old_codes
    in_new = code in new_codes
    mm = (row.get("legacy_mm_type") or row.get("new_mm_type") or "")
    if in_new and (row.get("new_mm_type","").lower().startswith("micro")):
        return "MICRO_OR_OUT_OF_SCOPE"
    if in_old and not in_new:
        return "LEGACY_CANDIDATE"
    if in_old and in_new and (row.get("legacy_title_n") != row.get("new_title_hit_n")):
        return "LEGACY_CANDIDATE"
    if (not in_old) and (not in_new):
        return "OTHER_CANDIDATE"
    # default: exists in new and is metro -> likely just not in MEFI coverage for other reasons
    return "METRO_BUT_NOT_IN_MEFI"

unknown_df["status"] = unknown_df.apply(classify, axis=1)

# ---------- step 3: propose current code for LEGACY_CANDIDATEs via title matching ----------
legacy_rows = unknown_df[unknown_df["status"]=="LEGACY_CANDIDATE"].copy()
choices = l_new["CBSA_TITLE_N"].dropna().unique().tolist()
def best_match(name_n):
    if not name_n: return (pd.NA, 0.0)
    m = process.extractOne(name_n, choices, scorer=fuzz.WRatio)
    if not m: return (pd.NA, 0.0)
    return m[0], float(m[1])

legacy_rows[["match_title_n","match_score"]] = legacy_rows["legacy_title_n"].apply(lambda s: pd.Series(best_match(s)))
title2code = dict(zip(l_new["CBSA_TITLE_N"], l_new["CBSA_CODE"]))
title2title = dict(zip(l_new["CBSA_TITLE_N"], l_new["CBSA_TITLE"]))
legacy_rows["current_cbsa_code"] = legacy_rows["match_title_n"].map(title2code)
legacy_rows["current_title"]     = legacy_rows["match_title_n"].map(title2title)

# keep suggestions with high confidence (e.g., >=95); others for manual review
suggest_hi = legacy_rows[legacy_rows["match_score"]>=95].copy()
suggest_lo = legacy_rows[(legacy_rows["match_score"]>0) & (legacy_rows["match_score"]<95)].copy()

# ---------- outputs ----------
print("SMART unique CBSA total:", len(smart_cbsa))
print("Unknown vs MEFI2017:", len(smart_unknown))
print(unknown_df["status"].value_counts(dropna=False))

print("\nHigh-confidence legacy→current suggestions (>=95):", len(suggest_hi))
display(suggest_hi[["legacy_cbsa_code","legacy_title","current_cbsa_code","current_title","match_score"]].head(20))

# Save mapping CSVs for review
suggest_hi.assign(method="fuzzy_title", note=">=95").to_csv("/content/cbsa_legacy_to_current_hi.csv", index=False)
suggest_lo.assign(method="fuzzy_title", note="90–95 or below").to_csv("/content/cbsa_legacy_to_current_review.csv", index=False)
unknown_df.to_csv("/content/cbsa_unknown_classified.csv", index=False)
````

````python
def collapse_to_cbsa(df):
    # Keep one row per CBSA code with a representative title/type/state
    # Pick the first non-null in each group; adjust as needed
    cols_keep = ["CBSA_CODE","CBSA_TITLE","MM_TYPE","STATE_NAME","CBSA_TITLE_N"]
    temp = df.copy()
    for c in cols_keep:
        if c not in temp.columns:
            temp[c] = pd.NA
    out = (temp
           .groupby("CBSA_CODE", as_index=False)
           .agg({
               "CBSA_TITLE": "first",
               "CBSA_TITLE_N": "first",
               "MM_TYPE": "first",
               "STATE_NAME": "first"
           }))
    return out

l_old_cbsa = collapse_to_cbsa(l_old)
l_new_cbsa = collapse_to_cbsa(l_new)
````

````python
import pandas as pd
import numpy as np
import re
import os

# ===== Helpers =====
def z5(x):
    if pd.isna(x): return pd.NA
    s = str(x).strip()
    if s == "" or s.lower() == "nan": return pd.NA
    if re.fullmatch(r"-?\d+(\.\d+)?", s):
        try: s = str(int(float(s)))
        except: pass
    return s.zfill(5)

def norm_name(s):
    if pd.isna(s): return ""
    s = str(s).lower()
    for ch in ["–","—","(",")",".",","]:
        s = s.replace(ch, " ")
    s = s.replace("-", " ")
    return " ".join(s.split())

# ===== Your final mapping (division→parent CBSA + legacy renames) =====
legacy_to_current = {
    # Metropolitan Division → Parent CBSA
    "14454": "14460", "15764": "14460", "40484": "14460",          # Boston
    "19124": "19100", "23104": "19100",                            # Dallas
    "47664": "19820",                                              # Detroit
    "35004": "35620", "35084": "35620", "35614": "35620",          # New York
    "15804": "37980", "33874": "37980",                            # Philadelphia
    # Legacy renames (old → current)
    "17200": "30150",  # Claremont–Lebanon → Lebanon–Claremont
    "19430": "19380",  # Dayton–Springfield → Dayton
}

# ===== Scope flags to show in the joined table =====
micros = {"10100","25740","33500","35820"}     # Micropolitan (no MEFI)
puerto_rico = {"10380","41980"}                # PR metros (MEFI typically excluded)

def add_scope_flags(df, code_col="CBSA_CODE"):
    df = df.copy()
    df[code_col] = df[code_col].astype(str).str.zfill(5)
    df["is_micro"] = df[code_col].isin(micros).astype(int)
    df["is_pr"]    = df[code_col].isin(puerto_rico).astype(int)
    def _note(code):
        if code in micros:      return "Micropolitan area — MEFI not available"
        if code in puerto_rico: return "Puerto Rico — MEFI typically excluded"
        return ""
    df["scope_note"] = df[code_col].map(_note)
    return df

# ===== 1) Load BRFSS 2017 =====
# UPDATE THIS PATH to where your SMART_2017_individual.csv actually lives in your Colab.
BRFSS_2017_PATH = "/content/drive/MyDrive/Saksham_MEFI/SMART_2017_individual.csv"  # e.g., "/content/SMART_2017_individual.csv" or your Drive path

BR2017 = pd.read_csv(BRFSS_2017_PATH, dtype=str, low_memory=False)
BR2017["CBSA_CODE"] = BR2017["CBSA_CODE"].apply(z5)
BR2017["CBSA_CODE_CANON"] = BR2017["CBSA_CODE"].replace(legacy_to_current)

# Flags & notes (original and canonical)
BR2017 = add_scope_flags(BR2017, code_col="CBSA_CODE")
tmp = add_scope_flags(BR2017.rename(columns={"CBSA_CODE_CANON":"CBSA_CODE_CANON_tmp"}), "CBSA_CODE_CANON_tmp")
BR2017["is_micro_canon"] = tmp["is_micro"]
BR2017["is_pr_canon"]    = tmp["is_pr"]
BR2017["scope_note_canon"] = tmp["scope_note"]
BR2017["scope_note"] = np.where(BR2017["scope_note_canon"].ne(""), BR2017["scope_note_canon"], BR2017["scope_note"])
BR2017.drop(columns=["scope_note_canon"], inplace=True)

# ===== 2) Load MEFI (scores panel) & prep 2017 slice =====
MEFI_XLSX = "/content/drive/MyDrive/Saksham_MEFI/MEFI+2022+dataset+for+researchers,+1-5-23 (4).xlsx"
scores = pd.read_excel(MEFI_XLSX, sheet_name="scores (panel)")
scores = scores.rename(columns={
    "CBSA Code": "CBSA_CODE",
    "Area 1: Govern-ment Spending": "Area1_GovtSpending",
    "Area 2: Taxes": "Area2_Taxes",
    "Area 3: Labor Market Regulation": "Area3_Labor",
    "Primary\nState": "Primary_State"
})
scores["CBSA_CODE"] = scores["CBSA_CODE"].apply(z5)
MEFI2017 = scores[scores["year"]==2017].copy()
MEFI2017["Metro_Area_Name"] = MEFI2017.get("Unnamed: 2", pd.Series([pd.NA]*len(MEFI2017)))
MEFI2017["Metro_Area_Name_N"] = MEFI2017["Metro_Area_Name"].apply(norm_name)

MEFI2017_ready = (MEFI2017
    .rename(columns={
        "Overall": "MEFI2017_overall",
        "Area1_GovtSpending": "MEFI2017_area1",
        "Area2_Taxes": "MEFI2017_area2",
        "Area3_Labor": "MEFI2017_area3"
    })
    [["CBSA_CODE","Metro_Area_Name","Metro_Area_Name_N",
      "MEFI2017_overall","MEFI2017_area1","MEFI2017_area2","MEFI2017_area3"]]
    .drop_duplicates(subset=["CBSA_CODE"])
)

# (Optional) Attach BEA population (2017) by normalized metro name
try:
    bea = pd.read_excel(MEFI_XLSX, sheet_name="BEA", header=[0,1])
    pop_col = next(col for col in bea.columns
                   if isinstance(col, tuple)
                   and str(col[0]).strip().lower().startswith("total population")
                   and str(col[1]).strip().startswith("2017"))
    name_col = next((col for col in bea.columns if isinstance(col, tuple) and "metro area" in str(col[0]).lower()),
                    next((col for col in bea.columns if isinstance(col, str) and "metro area" in col.lower()), None))
    if name_col is not None:
        BEA2017 = bea[[name_col, pop_col]].copy()
        BEA2017.columns = ["Metro_Area_Name","BEA_pop_2017"]
        BEA2017 = BEA2017[pd.to_numeric(BEA2017["BEA_pop_2017"], errors="coerce").notna()].copy()
        BEA2017["BEA_pop_2017"] = BEA2017["BEA_pop_2017"].astype(float)
        BEA2017["Metro_Area_Name_N"] = BEA2017["Metro_Area_Name"].apply(norm_name)
        MEFI2017_ready = MEFI2017_ready.merge(BEA2017[["Metro_Area_Name_N","BEA_pop_2017"]],
                                              on="Metro_Area_Name_N", how="left")
except StopIteration:
    # If BEA column detection fails, just skip population for now
    pass

# ===== 3) Merge BRFSS 2017 ↔ MEFI 2017 (CBSA_CODE_CANON ↔ CBSA_CODE) =====
merged2017 = BR2017.merge(
    MEFI2017_ready.drop(columns=["Metro_Area_Name_N"]),
    left_on="CBSA_CODE_CANON", right_on="CBSA_CODE", how="left"
)

mefi_cols = ["MEFI2017_overall","MEFI2017_area1","MEFI2017_area2","MEFI2017_area3"]
for c in mefi_cols:
    if c not in merged2017.columns:
        merged2017[c] = np.nan

merged2017["mefi_joined"] = merged2017[mefi_cols].notna().any(axis=1).astype(int)

def _join_note(row):
    if row["mefi_joined"] == 1:
        return "MEFI attached"
    if row.get("scope_note"):
        return row["scope_note"]
    return "No MEFI match (unexpected) — check mapping"

def _status(row):
    if row["mefi_joined"] == 1:
        return "joined"
    if row.get("is_micro_canon",0)==1 or row.get("is_pr_canon",0)==1 or row.get("is_micro",0)==1 or row.get("is_pr",0)==1:
        return "out_of_scope"
    return "unmatched"

merged2017["mefi_join_note"] = merged2017.apply(_join_note, axis=1)
merged2017["join_status"] = merged2017.apply(_status, axis=1)

# ===== 4) Show 100 random samples =====
view_cols = [
    "CBSA_CODE","CBSA_CODE_CANON","join_status","mefi_join_note",
    "is_micro","is_pr","is_micro_canon","is_pr_canon","scope_note",
    "MEFI2017_overall","MEFI2017_area1","MEFI2017_area2","MEFI2017_area3","BEA_pop_2017",
    "Metro_Area_Name"
]
view_cols = [c for c in view_cols if c in merged2017.columns]
n = min(100, len(merged2017))
display(merged2017.sample(n=n, random_state=42)[view_cols])

# Optional quick summary
print({
    "rows_total": len(merged2017),
    "unique_CBSA_original": merged2017["CBSA_CODE"].nunique(),
    "unique_CBSA_canon": merged2017["CBSA_CODE_CANON"].nunique(),
    "joined_rows": int((merged2017["join_status"]=="joined").sum()),
    "out_of_scope_rows": int((merged2017["join_status"]=="out_of_scope").sum()),
    "unmatched_rows": int((merged2017["join_status"]=="unmatched").sum()),
})
````

````python
import pandas as pd
import numpy as np
import re

# --------- Helpers ----------
def z5(x):
    if pd.isna(x): return pd.NA
    s = str(x).strip()
    if s == "" or s.lower() == "nan": return pd.NA
    if re.fullmatch(r"-?\d+(\.\d+)?", s):
        try: s = str(int(float(s)))
        except: pass
    return s.zfill(5)

def norm_name(s):
    if pd.isna(s): return ""
    s = str(s).lower()
    for ch in ["–","—","(",")",".",","]:
        s = s.replace(ch, " ")
    s = s.replace("-", " ")
    return " ".join(s.split())

def add_scope_flags(df, code_col="CBSA_CODE_CANON"):
    micros = {"10100","25740","33500","35820"}
    puerto_rico = {"10380","41980"}
    df["is_micro"] = df[code_col].isin(micros).astype(int)
    df["is_pr"]    = df[code_col].isin(puerto_rico).astype(int)
    def _note(code):
        if code in micros:      return "Micropolitan area — MEFI not available"
        if code in puerto_rico: return "Puerto Rico — MEFI typically excluded"
        return ""
    df["scope_note"] = df[code_col].map(_note)
    return df

# --------- Inputs (update paths) ----------
SMART2017 = pd.read_csv("/content/drive/MyDrive/Saksham_MEFI/SMART_2017_individual.csv", dtype=str, low_memory=False)
SMART2022 = pd.read_csv("/content/drive/MyDrive/Saksham_MEFI/SMART_2022_individual.csv", dtype=str, low_memory=False)
MEFI_XLSX = "/content/drive/MyDrive/Saksham_MEFI/MEFI+2022+dataset+for+researchers,+1-5-23 (4).xlsx"
LIST1_2023 = "/content/drive/MyDrive/Saksham_MEFI/list1_2023.xlsx"

# --------- Build division → parent CBSA from List 1 (2023) ----------
l1 = pd.read_excel(LIST1_2023, sheet_name="List 1", dtype=str, header=2)
l1.rename(columns=lambda c: c.strip(), inplace=True)
if "CBSA Code" in l1.columns: l1["CBSA Code"] = l1["CBSA Code"].apply(z5)
if "Metropolitan Division Code" in l1.columns: l1["Metropolitan Division Code"] = l1["Metropolitan Division Code"].apply(z5)

div2cbsa = {}
if {"Metropolitan Division Code","CBSA Code"}.issubset(l1.columns):
    div2cbsa = (l1[["Metropolitan Division Code","CBSA Code"]]
                .dropna().drop_duplicates()
                .set_index("Metropolitan Division Code")["CBSA Code"].to_dict())

# --------- Legacy→current mapping you confirmed ----------
legacy_to_current = {
    # Metropolitan Division → Parent CBSA
    "14454":"14460","15764":"14460","40484":"14460",     # Boston
    "19124":"19100","23104":"19100",                     # Dallas
    "47664":"19820",                                     # Detroit
    "35004":"35620","35084":"35620","35614":"35620",     # New York
    "15804":"37980","33874":"37980",                     # Philadelphia
    # Legacy renames
    "17200":"30150",                                     # Claremont–Lebanon → Lebanon–Claremont
    "19430":"19380",                                     # Dayton–Springfield → Dayton
}

# --------- Prepare BRFSS (function) ----------
def prep_brfss(df):
    df = df.copy()
    assert "CBSA_CODE" in df.columns, "CBSA_CODE missing in BRFSS file."
    df["CBSA_CODE_ORIG"]   = df["CBSA_CODE"].apply(z5)
    df["CBSA_CODE_ROLLED"] = df["CBSA_CODE_ORIG"].map(div2cbsa).fillna(df["CBSA_CODE_ORIG"])
    df["CBSA_CODE_CANON"]  = df["CBSA_CODE_ROLLED"].replace(legacy_to_current)
    df = add_scope_flags(df, "CBSA_CODE_CANON")
    # Drop the original CBSA_CODE column before merging to prevent overlap later
    if "CBSA_CODE" in df.columns:
        df = df.drop(columns=["CBSA_CODE"])
    return df

B17 = prep_brfss(SMART2017)
B22 = prep_brfss(SMART2022)

# --------- Load MEFI “scores (panel)” (Overall + Area1–3) @ 2017 ----------
scores = pd.read_excel(MEFI_XLSX, sheet_name="scores (panel)", header=0) # Assuming header is on row 0
scores = scores.rename(columns={
    "CBSA Code":"CBSA_CODE",
    "Area 1: Govern-ment Spending":"Area1_GovtSpending",
    "Area 2: Taxes":"Area2_Taxes",
    "Area 3: Labor Market Regulation":"Area3_Labor",
    "Primary\nState":"Primary_State"
})
scores["CBSA_CODE"] = scores["CBSA_CODE"].apply(z5)
MEFI2017 = scores[scores["year"]==2017].copy()
MEFI2017["Metro_Area_Name"] = MEFI2017.get("Unnamed: 2", pd.Series([pd.NA]*len(MEFI2017)))
MEFI2017["Metro_Area_Name_N"] = MEFI2017["Metro_Area_Name"].apply(norm_name)
MEFI2017_main = (MEFI2017
    .rename(columns={
        "Overall":"MEFI2017_overall",
        "Area1_GovtSpending":"MEFI2017_area1",
        "Area2_Taxes":"MEFI2017_area2",
        "Area3_Labor":"MEFI2017_area3"
    })
    [["CBSA_CODE","Metro_Area_Name","Metro_Area_Name_N",
      "MEFI2017_overall","MEFI2017_area1","MEFI2017_area2","MEFI2017_area3"]]
    .drop_duplicates(subset=["CBSA_CODE"])
)

# --------- Pull 1A–3C (score) for 2017 from the 9 sheets (wide → tidy) ----------
def pull_component(mefi_path, sheet_name, comp_label):
    # These sheets have header information in the first few rows, but the actual column names we want
    # for the scores (like '1A (score)') are on a specific row, and the years are sometimes
    # in the row below that. Let's inspect the sheets to find the correct header row.
    # Based on the output of cell cf6b3879, the column headers are typically in row 0 or 1.
    # The actual data seems to start from row 5 onwards for the component sheets.
    # Let's assume the column names we want are in the first row (index 0) and the year
    # information is sometimes in the second row (index 1), but the actual score for 2017
    # is in a column labeled '2017' under the score column.

    # Try reading with header=0 first and see if we can find the 2017 column under the score column.
    try:
        df = pd.read_excel(mefi_path, sheet_name=sheet_name, header=0)
        # Flatten MultiIndex if present
        df.columns = pd.Index([ " ".join(map(str,c)) if isinstance(c, tuple) else str(c) for c in df.columns ])

        cbsa_col = None
        for cand in ["CBSA Code","CBSA code","CBSA CODE","CBSA"]:
            if cand in df.columns: cbsa_col = cand; break
        if cbsa_col is None:
             # Fallback to the second column if no CBSA column is found by name
            if len(df.columns) > 1:
                cbsa_col = df.columns[1]
            else:
                return pd.DataFrame(columns=["CBSA_CODE", f"MEFI2017_{comp_label}"]) # Cannot find CBSA column

        df["CBSA_CODE"] = df[cbsa_col].apply(z5)

        # Find the 2017 score column
        year_col = None
        # Search for a column that contains the component label and "score"
        score_col_candidate = [c for c in df.columns if comp_label in c and "(score)" in c]
        if score_col_candidate:
             # If found, look for a sub-column named '2017' under this
             if f"{score_col_candidate[0]} 2017" in df.columns:
                 year_col = f"{score_col_candidate[0]} 2017"
             else:
                 # Fallback: look for a column named exactly '2017'
                 maybe_year = [c for c in df.columns if c.strip() == "2017"]
                 if maybe_year:
                     year_col = maybe_year[0]
        else:
            # If no column with (score) is found, try looking for a column named '2017' directly
            maybe_year = [c for c in df.columns if c.strip() == "2017"]
            if maybe_year:
                year_col = maybe_year[0]

        out = df[["CBSA_CODE"] + ([year_col] if year_col else [])].copy()
        if year_col:
            out.rename(columns={year_col: f"MEFI2017_{comp_label}"}, inplace=True)
        else:
            out[f"MEFI2017_{comp_label}"] = np.nan
        return out[["CBSA_CODE", f"MEFI2017_{comp_label}"]]

    except Exception as e:
        print(f"Error processing sheet {sheet_name}: {e}")
        return pd.DataFrame(columns=["CBSA_CODE", f"MEFI2017_{comp_label}"])


components = []
components += [pull_component(MEFI_XLSX, "1A", "1A")]
components += [pull_component(MEFI_XLSX, "1B", "1B")]
components += [pull_component(MEFI_XLSX, "1C", "1C")]
components += [pull_component(MEFI_XLSX, "2A", "2A")]
components += [pull_component(MEFI_XLSX, "2B", "2B")]
components += [pull_component(MEFI_XLSX, "2C", "2C")]
components += [pull_component(MEFI_XLSX, "3A", "3A")]
components += [pull_component(MEFI_XLSX, "3B", "3B")]
components += [pull_component(MEFI_XLSX, "3C", "3C")]

MEFI2017_comps = components[0]
for t in components[1:]:
    MEFI2017_comps = MEFI2017_comps.merge(t, on="CBSA_CODE", how="outer")

# --------- BEA population 2017 ----------
# The BEA sheet has a two-row header, so we specify header=[0, 1]
try:
    bea = pd.read_excel(MEFI_XLSX, sheet_name="BEA", header=[0,1])
    pop_col = None
    for col in bea.columns:
        if isinstance(col, tuple):
            if str(col[0]).strip().lower().startswith("total population") and str(col[1]).strip().startswith("2017"):
                pop_col = col; break
    name_col = None
    for col in bea.columns:
        if isinstance(col, tuple) and "metro area" in str(col[0]).lower():
            name_col = col; break
    if name_col is None:
        for col in bea.columns:
            if isinstance(col, str) and "metro area" in col.lower():
                name_col = col; break

    MEFI2017_bea = pd.DataFrame(columns=["CBSA_CODE","BEA_pop_2017"])
    if pop_col is not None and name_col is not None:
        BEA2017 = bea[[name_col, pop_col]].copy()
        BEA2017.columns = ["Metro_Area_Name","BEA_pop_2017"]
        BEA2017 = BEA2017[pd.to_numeric(BEA2017["BEA_pop_2017"], errors="coerce").notna()].copy()
        BEA2017["BEA_pop_2017"] = BEA2017["BEA_pop_2017"].astype(float)
        BEA2017["Metro_Area_Name_N"] = BEA2017["Metro_Area_Name"].apply(norm_name)
        # Merge to MEFI2017_main by normalized name to pick the population
        MEFI2017_main = MEFI2017_main.merge(
            BEA2017[["Metro_Area_Name_N","BEA_pop_2017"]],
            on="Metro_Area_Name_N", how="left"
        )

except Exception as e:
    print(f"Error processing BEA sheet: {e}")
    # If BEA processing fails, just skip population for now


# --------- Assemble the MEFI 2017 bundle ----------
# Merge MEFI2017_main and MEFI2017_comps by CBSA_CODE
MEFI2017_all = (MEFI2017_main
    .merge(MEFI2017_comps, on="CBSA_CODE", how="left")
    .drop(columns=["Metro_Area_Name_N"])
    .drop_duplicates(subset=["CBSA_CODE"])
)

# --------- Merge BRFSS 2017 ↔ MEFI 2017 ----------
def merge_brfss_mefi(brfss_df, mefi_df, year_label="2017"):
    # Join BRFSS on CBSA_CODE_CANON and MEFI on CBSA_CODE
    out = brfss_df.merge(mefi_df, left_on="CBSA_CODE_CANON", right_on="CBSA_CODE", how="left") # Removed suffixes as CBSA_CODE_CANON is unique
    # Status + note
    mefi_core = [f"MEFI{year_label}_overall", f"MEFI{year_label}_area1", f"MEFI{year_label}_area2", f"MEFI{year_label}_area3"]
    # Ensure mefi_core columns exist before checking notna
    mefi_core_existing = [c for c in mefi_core if c in out.columns]
    out["mefi_joined"] = out[mefi_core_existing].notna().any(axis=1).astype(int)
    out["join_status"] = np.where(out["mefi_joined"].eq(1), "joined",
                           np.where((out["is_micro"].eq(1) | out["is_pr"].eq(1)), "out_of_scope", "unmatched"))
    out["mefi_join_note"] = np.where(out["mefi_joined"].eq(1), "MEFI attached",
                               np.where(out["scope_note"].ne(""), out["scope_note"], "No MEFI match — check mapping"))
    return out

BRFSS2017_MEFI = merge_brfss_mefi(B17, MEFI2017_all, "2017")

# --------- For 2022, spec says use MEFI 2017 (lag experiment) ----------
BRFSS2022_MEFI = merge_brfss_mefi(B22, MEFI2017_all, "2017")  # yes, 2017 MEFI for 2022 per instruction
````

````python
def pull_component_by_name(mefi_path, sheet_name, comp_label, year=2017):
    # Read as two-row header
    df = pd.read_excel(mefi_path, sheet_name=sheet_name, header=[0,1])
    # Find Metro Area column
    metro_col = None
    for col in df.columns:
        if (isinstance(col, tuple) and "metro area" in str(col[0]).lower()) or \
           (isinstance(col, str) and "metro area" in col.lower()):
            metro_col = col
            break
    if metro_col is None:
        raise KeyError(f"Could not find 'Metro Area' in {sheet_name}")

    # Candidate year labels
    year_labels = [year, f"{year}.0", str(year)]
    # Locate ('1A (score)', 2017) style column
    comp_col = None
    for col in df.columns:
        top = col[0] if isinstance(col, tuple) else col
        sub = col[1] if isinstance(col, tuple) else None
        if comp_label in str(top) and "(score)" in str(top):
            if (isinstance(col, tuple) and (sub in year_labels)) or \
               (not isinstance(col, tuple) and str(top).strip()==str(year)):
                comp_col = col
                break
    if comp_col is None:
        # fallback: scan for any column whose top contains comp_label and subheader equals one of year_labels
        for col in df.columns:
            if isinstance(col, tuple):
                if comp_label in str(col[0]) and "(score)" in str(col[0]) and str(col[1]) in map(str, year_labels):
                    comp_col = col
                    break
    # Build tidy
    out = df[[metro_col]].copy()
    out.columns = ["Metro_Area_Name"]
    out[f"MEFI{year}_{comp_label}"] = df[comp_col] if comp_col in df.columns else np.nan
    # normalize name for joining
    out["Metro_Area_Name_N"] = out["Metro_Area_Name"].apply(norm_name)
    # keep one row per metro
    out = (out.groupby("Metro_Area_Name_N", as_index=False)
              .agg({f"MEFI{year}_{comp_label}":"first"}))
    return out
````

````python
import pandas as pd
import numpy as np
import re
import json

# ========= Helpers =========
def z5(x):
    """Pad CBSA codes to 5 digits with leading zeros"""
    if pd.isna(x): return pd.NA
    s = str(x).strip()
    if s == "" or s.lower() == "nan": return pd.NA
    if re.fullmatch(r"-?\d+(\.\d+)?", s):
        try: s = str(int(float(s)))
        except: pass
    return s.zfill(5)

def norm_name(s):
    """Normalize metro area names for matching"""
    if pd.isna(s): return ""
    s = str(s).lower()
    for ch in ["–","—","(",")",".",","]:
        s = s.replace(ch, " ")
    s = s.replace("-", " ")
    return " ".join(s.split())

def add_scope_flags(df, code_col="CBSA_CODE_CANON"):
    """Add flags for micropolitan areas and Puerto Rico"""
    micros = {"10100","25740","33500","35820"}
    puerto_rico = {"10380","41980"}
    df["is_micro"] = df[code_col].isin(micros).astype(int)
    df["is_pr"]    = df[code_col].isin(puerto_rico).astype(int)
    def _note(code):
        if code in micros:      return "Micropolitan area — MEFI not available"
        if code in puerto_rico: return "Puerto Rico — MEFI typically excluded"
        return ""
    df["scope_note"] = df[code_col].map(_note)
    return df

SUMMARY_ROWS = {"min","max","average","median"}

def pull_component_by_name(mefi_path, sheet_name, comp_label, year=2017):
    """
    Extract component scores from MEFI sheets with two-row headers.
    """
    print(f"  Processing sheet: {sheet_name}")

    # Read with 2-row header
    df = pd.read_excel(mefi_path, sheet_name=sheet_name, header=[0,1])

    # Debug: print first few column names to understand structure
    print(f"    First 10 columns: {df.columns[:10].tolist()}")

    # Find Metro Area column - check all parts of tuple column names
    metro_col = None
    for col in df.columns:
        if isinstance(col, tuple):
            # Check both levels of the tuple
            col_combined = ' '.join([str(c).lower() for c in col])
            # Also check individual parts
            for part in col:
                part_str = str(part).lower()
                if "metro area" in part_str or "metro area" in col_combined:
                    metro_col = col
                    print(f"    Found Metro Area column: {col}")
                    break
            if metro_col:
                break
        else:
            # Single-level column name
            col_str = str(col).lower()
            if "metro area" in col_str:
                metro_col = col
                print(f"    Found Metro Area column: {col}")
                break

    if metro_col is None:
        print(f"    Warning: No 'Metro Area' column found in {sheet_name}")
        print(f"    Available columns: {df.columns.tolist()[:15]}")
        return pd.DataFrame({"Metro_Area_Name_N": [], f"MEFI{year}_{comp_label}": []})

    # Find component score column: e.g., ('1A (score)', '2017')
    comp_col = None
    year_labels = {str(year), f"{year}.0", year, float(year)}

    for col in df.columns:
        if isinstance(col, tuple) and len(col) >= 2:
            top, sub = str(col[0]), str(col[1])
            # Match component label with "(score)" and year
            if comp_label.lower() in top.lower() and "(score)" in top.lower():
                # Check if year matches
                if sub in year_labels or str(sub) in year_labels:
                    comp_col = col
                    print(f"    Found score column: {col}")
                    break
                # Also try numeric comparison
                try:
                    if float(sub) == float(year):
                        comp_col = col
                        print(f"    Found score column: {col}")
                        break
                except:
                    pass

    if comp_col is None:
        print(f"    Warning: No score column found for {comp_label} in {sheet_name}")
        print(f"    Looking for pattern: {comp_label}* with year {year}")
        return pd.DataFrame({"Metro_Area_Name_N": [], f"MEFI{year}_{comp_label}": []})

    # Extract the two columns
    out = df[[metro_col, comp_col]].copy()
    out.columns = ["Metro_Area_Name", f"MEFI{year}_{comp_label}"]

    # Drop summary rows
    mask_summary = out["Metro_Area_Name"].astype(str).str.strip().str.lower().isin(SUMMARY_ROWS)
    out = out[~mask_summary].copy()

    # Drop rows where metro name is NaN or empty
    out = out[out["Metro_Area_Name"].notna()].copy()
    out = out[out["Metro_Area_Name"].astype(str).str.strip() != ""].copy()

    # Normalize name for joining
    out["Metro_Area_Name_N"] = out["Metro_Area_Name"].apply(norm_name)

    # Drop empty normalized names
    out = out[out["Metro_Area_Name_N"] != ""].copy()

    # Deduplicate
    out = (out.sort_values("Metro_Area_Name_N")
             .drop_duplicates(subset=["Metro_Area_Name_N"], keep="first"))

    print(f"    Extracted {len(out)} rows")
    return out[["Metro_Area_Name_N", f"MEFI{year}_{comp_label}"]]

# ========= File Paths (UPDATE THESE) =========
SMART2017 = pd.read_csv("/content/drive/MyDrive/Saksham_MEFI/SMART_2017_individual.csv",
                        dtype=str, low_memory=False)
SMART2022 = pd.read_csv("/content/drive/MyDrive/Saksham_MEFI/SMART_2022_individual.csv",
                        dtype=str, low_memory=False)
MEFI_XLSX  = "/content/drive/MyDrive/Saksham_MEFI/MEFI+2022+dataset+for+researchers,+1-5-23 (4).xlsx"
LIST1_2023 = "/content/drive/MyDrive/Saksham_MEFI/list1_2023.xlsx"

print("=" * 60)
print("STEP 1: Processing OMB List 1 for Metro Division mappings")
print("=" * 60)

# ========= Division → parent CBSA (2023 List 1) =========
l1 = pd.read_excel(LIST1_2023, sheet_name="List 1", dtype=str, header=2)
l1.rename(columns=lambda c: c.strip(), inplace=True)

if "CBSA Code" in l1.columns:
    l1["CBSA Code"] = l1["CBSA Code"].apply(z5)
if "Metropolitan Division Code" in l1.columns:
    l1["Metropolitan Division Code"] = l1["Metropolitan Division Code"].apply(z5)

div2cbsa = {}
if {"Metropolitan Division Code","CBSA Code"}.issubset(l1.columns):
    div2cbsa = (l1[["Metropolitan Division Code","CBSA Code"]]
                .dropna().drop_duplicates()
                .set_index("Metropolitan Division Code")["CBSA Code"].to_dict())

print(f"  Found {len(div2cbsa)} Metro Division → CBSA mappings")

# ========= Legacy → current mapping =========
legacy_to_current = {
    # Metro Division → Parent CBSA
    "14454":"14460","15764":"14460","40484":"14460",     # Boston
    "19124":"19100","23104":"19100",                     # Dallas
    "47664":"19820",                                     # Detroit
    "35004":"35620","35084":"35620","35614":"35620",     # New York
    "15804":"37980","33874":"37980",                     # Philadelphia
    # Legacy renames
    "17200":"30150",                                     # Claremont–Lebanon → Lebanon–Claremont
    "19430":"19380",                                     # Dayton–Springfield → Dayton
}

def prep_brfss(df):
    """Prepare BRFSS data with CBSA code standardization"""
    df = df.copy()
    assert "CBSA_CODE" in df.columns, "CBSA_CODE missing in BRFSS file."
    df["CBSA_CODE_ORIG"]   = df["CBSA_CODE"].apply(z5)
    df["CBSA_CODE_ROLLED"] = df["CBSA_CODE_ORIG"].map(div2cbsa).fillna(df["CBSA_CODE_ORIG"])
    df["CBSA_CODE_CANON"]  = df["CBSA_CODE_ROLLED"].replace(legacy_to_current)
    df = add_scope_flags(df, "CBSA_CODE_CANON")
    df = df.drop(columns=["CBSA_CODE"], errors="ignore")
    return df

print("\n" + "=" * 60)
print("STEP 2: Preparing BRFSS 2017 and 2022 data")
print("=" * 60)

B17 = prep_brfss(SMART2017)
B22 = prep_brfss(SMART2022)

print(f"  BRFSS 2017: {len(B17):,} rows")
print(f"  BRFSS 2022: {len(B22):,} rows")

print("\n" + "=" * 60)
print("STEP 3: Reading MEFI panel scores (Overall + Area1-3)")
print("=" * 60)

# ========= MEFI panel (Overall + Area1–3) @ 2017 =========
scores = pd.read_excel(MEFI_XLSX, sheet_name="scores (panel)", header=0)
scores = scores.rename(columns={
    "CBSA Code":"CBSA_CODE",
    "Area 1: Govern-ment Spending":"Area1_GovtSpending",
    "Area 2: Taxes":"Area2_Taxes",
    "Area 3: Labor Market Regulation":"Area3_Labor",
    "Primary\nState":"Primary_State"
})
scores["CBSA_CODE"] = scores["CBSA_CODE"].apply(z5)

MEFI2017 = scores[scores["year"]==2017].copy()
MEFI2017["Metro_Area_Name"] = MEFI2017.get("Unnamed: 2", pd.Series([pd.NA]*len(MEFI2017)))
MEFI2017["Metro_Area_Name_N"] = MEFI2017["Metro_Area_Name"].apply(norm_name)

MEFI2017_main = (MEFI2017
    .rename(columns={
        "Overall":"MEFI2017_overall",
        "Area1_GovtSpending":"MEFI2017_area1",
        "Area2_Taxes":"MEFI2017_area2",
        "Area3_Labor":"MEFI2017_area3"
    })
    [["CBSA_CODE","Metro_Area_Name","Metro_Area_Name_N",
      "MEFI2017_overall","MEFI2017_area1","MEFI2017_area2","MEFI2017_area3"]]
    .drop_duplicates(subset=["CBSA_CODE"])
)

print(f"  Extracted {len(MEFI2017_main)} metro areas from panel data")

print("\n" + "=" * 60)
print("STEP 4: Extracting component scores (1A-3C) from individual sheets")
print("=" * 60)

# ========= Components 1A–3C by Metro name (two-row headers) =========
parts = []
for sheet, label in [("1A","1A"),("1B","1B"),("1C","1C"),
                     ("2A","2A"),("2B","2B"),("2C","2C"),
                     ("3A","3A"),("3B","3B"),("3C","3C")]:
    parts.append(pull_component_by_name(MEFI_XLSX, sheet, label, year=2017))

MEFI2017_comps = parts[0]
for t in parts[1:]:
    MEFI2017_comps = MEFI2017_comps.merge(t, on="Metro_Area_Name_N", how="outer")

print(f"\n  Combined components: {len(MEFI2017_comps)} unique metro areas")

print("\n" + "=" * 60)
print("STEP 5: Extracting BEA population data")
print("=" * 60)

# ========= BEA population 2017 (two-row header; join by Metro name) =========
try:
    bea = pd.read_excel(MEFI_XLSX, sheet_name="BEA", header=[0,1])

    print(f"  BEA sheet columns (first 10): {bea.columns[:10].tolist()}")

    # Find population column for 2017
    pop_col = None
    for col in bea.columns:
        if isinstance(col, tuple):
            col_combined = ' '.join([str(c).lower() for c in col])
            if "total population" in col_combined or "population" in str(col[0]).lower():
                # Check if 2017 is in the column
                if "2017" in str(col[1]) or "2017" in col_combined:
                    pop_col = col
                    print(f"  Found population column: {col}")
                    break

    # Find metro area name column
    name_col = None
    for col in bea.columns:
        if isinstance(col, tuple):
            col_combined = ' '.join([str(c).lower() for c in col])
            for part in col:
                if "metro area" in str(part).lower() or "metro area" in col_combined:
                    name_col = col
                    print(f"  Found metro name column: {col}")
                    break
            if name_col:
                break
        else:
            if "metro area" in str(col).lower():
                name_col = col
                print(f"  Found metro name column: {col}")
                break

    if pop_col is not None and name_col is not None:
        BEA2017 = bea[[name_col, pop_col]].copy()
        BEA2017.columns = ["Metro_Area_Name","BEA_pop_2017"]
        BEA2017 = BEA2017[pd.to_numeric(BEA2017["BEA_pop_2017"], errors="coerce").notna()].copy()
        BEA2017["BEA_pop_2017"] = BEA2017["BEA_pop_2017"].astype(float)
        BEA2017["Metro_Area_Name_N"] = BEA2017["Metro_Area_Name"].apply(norm_name)
        MEFI2017_main = MEFI2017_main.merge(BEA2017[["Metro_Area_Name_N","BEA_pop_2017"]],
                                            on="Metro_Area_Name_N", how="left")
        print(f"  Added population data for {BEA2017['BEA_pop_2017'].notna().sum()} metro areas")
    else:
        print("  Warning: Could not find population columns in BEA sheet")
except Exception as e:
    print(f"  Warning: Could not extract BEA data: {e}")

print("\n" + "=" * 60)
print("STEP 6: Assembling complete MEFI 2017 dataset")
print("=" * 60)

# ========= Assemble MEFI 2017 bundle =========
MEFI2017_all = (MEFI2017_main
                .merge(MEFI2017_comps, on="Metro_Area_Name_N", how="left")
                .drop(columns=["Metro_Area_Name_N"])
                .drop_duplicates(subset=["CBSA_CODE"])
)

print(f"  Complete MEFI dataset: {len(MEFI2017_all)} metro areas")
print(f"  Columns: {list(MEFI2017_all.columns)}")

print("\n" + "=" * 60)
print("STEP 7: Merging BRFSS with MEFI data")
print("=" * 60)

# ========= Merge function (used for 2017 & 2022) =========
def merge_brfss_mefi(brfss_df, mefi_df, year_label="2017"):
    """Merge BRFSS data with MEFI scores and add join status flags"""
    out = brfss_df.merge(mefi_df, left_on="CBSA_CODE_CANON", right_on="CBSA_CODE", how="left")
    mefi_core = [f"MEFI{year_label}_overall", f"MEFI{year_label}_area1",
                 f"MEFI{year_label}_area2", f"MEFI{year_label}_area3"]
    mefi_core_existing = [c for c in mefi_core if c in out.columns]
    out["mefi_joined"] = out[mefi_core_existing].notna().any(axis=1).astype(int)
    out["join_status"] = np.where(out["mefi_joined"].eq(1), "joined",
                           np.where((out["is_micro"].eq(1) | out["is_pr"].eq(1)), "out_of_scope", "unmatched"))
    out["mefi_join_note"] = np.where(out["mefi_joined"].eq(1), "MEFI attached",
                               np.where(out["scope_note"].ne(""), out["scope_note"], "No MEFI match — check mapping"))
    return out

# ========= Final merges =========
BRFSS2017_MEFI = merge_brfss_mefi(B17, MEFI2017_all, "2017")
BRFSS2022_MEFI = merge_brfss_mefi(B22, MEFI2017_all, "2017")  # per spec: 2022 uses MEFI 2017

print(f"  BRFSS 2017 + MEFI: {len(BRFSS2017_MEFI):,} rows")
print(f"    - Joined: {(BRFSS2017_MEFI['join_status']=='joined').sum():,}")
print(f"    - Out of scope: {(BRFSS2017_MEFI['join_status']=='out_of_scope').sum():,}")
print(f"    - Unmatched: {(BRFSS2017_MEFI['join_status']=='unmatched').sum():,}")

print(f"\n  BRFSS 2022 + MEFI: {len(BRFSS2022_MEFI):,} rows")
print(f"    - Joined: {(BRFSS2022_MEFI['join_status']=='joined').sum():,}")
print(f"    - Out of scope: {(BRFSS2022_MEFI['join_status']=='out_of_scope').sum():,}")
print(f"    - Unmatched: {(BRFSS2022_MEFI['join_status']=='unmatched').sum():,}")

print("\n" + "=" * 60)
print("STEP 8: Saving output files (Parquet + CSV)")
print("=" * 60)

# ========= Save final files - BOTH PARQUET AND CSV =========
out2017_parquet = "/content/drive/MyDrive/Saksham_MEFI/BRFSS2017_MEFI.parquet"
out2022_parquet = "/content/drive/MyDrive/Saksham_MEFI/BRFSS2022_MEFI.parquet"
out2017_csv = "/content/drive/MyDrive/Saksham_MEFI/BRFSS2017_MEFI.csv"
out2022_csv = "/content/drive/MyDrive/Saksham_MEFI/BRFSS2022_MEFI.csv"

# Save Parquet files
BRFSS2017_MEFI.to_parquet(out2017_parquet, index=False)
BRFSS2022_MEFI.to_parquet(out2022_parquet, index=False)

# Save CSV files
BRFSS2017_MEFI.to_csv(out2017_csv, index=False)
BRFSS2022_MEFI.to_csv(out2022_csv, index=False)

print(f"  ✓ Saved Parquet: {out2017_parquet}")
print(f"  ✓ Saved CSV: {out2017_csv}")
print(f"  ✓ Saved Parquet: {out2022_parquet}")
print(f"  ✓ Saved CSV: {out2022_csv}")

# Column dictionary for documentation
col_dict = {
  "join_keys": ["CBSA_CODE_ORIG","CBSA_CODE_ROLLED","CBSA_CODE_CANON","CBSA_CODE (from MEFI bundle)"],
  "flags": ["is_micro","is_pr","scope_note","mefi_joined","join_status","mefi_join_note"],
  "mefi_2017_core": ["MEFI2017_overall","MEFI2017_area1","MEFI2017_area2","MEFI2017_area3","BEA_pop_2017"],
  "mefi_2017_components": ["MEFI2017_1A","MEFI2017_1B","MEFI2017_1C",
                           "MEFI2017_2A","MEFI2017_2B","MEFI2017_2C",
                           "MEFI2017_3A","MEFI2017_3B","MEFI2017_3C"]
}

dict_path = "/content/drive/MyDrive/Saksham_MEFI/MEFI_columns_dictionary.json"
with open(dict_path, "w") as f:
    json.dump(col_dict, f, indent=2)

print(f"  ✓ Saved: {dict_path}")

print("\n" + "=" * 60)
print("PROCESSING COMPLETE!")
print("=" * 60)
print("\nOutput files:")
print("  - BRFSS2017_MEFI.parquet & .csv")
print("  - BRFSS2022_MEFI.parquet & .csv")
print("  - MEFI_columns_dictionary.json")
````

````python
"""
OPTION 1: MYSTERY CODE INVESTIGATION
Purpose: Identify what the 4 mystery codes are and how to fix them

This script searches OMB List 1 and MEFI files to determine:
- What are codes 47894, 35154, 47980, 43524, 39100?
- Are they Metropolitan Divisions? If so, what's the parent MSA?
- Are they legacy codes that were renamed?
- Should they be excluded?
"""

import pandas as pd
import numpy as np
import json

print("=" * 80)
print("MYSTERY CODE INVESTIGATION")
print("=" * 80)

# ============================================================================
# CONFIGURATION
# ============================================================================

# Mystery codes from discovery (causing most unmatched rows)
MYSTERY_CODES = {
    "47894": "8,542 rows unmatched (2017) + 8,759 rows (2022) = DC MDiv suspected",
    "43524": "2,442 rows unmatched (2017) = Silver Spring suspected",
    "35154": "2,235 rows unmatched (2022) = Unknown MDiv",
    "47980": "1,036 rows unmatched (2022) = Unknown",
    "39100": "517 rows unmatched (2022) = Poughkeepsie suspected",
    "30100": "2,066 rows unmatched (2022) = Lebanon area",
    "30150": "1,763 rows unmatched (2017) = Lebanon area",
}

# File paths (update if needed)
OMB_OLD = "/content/drive/MyDrive/Saksham_MEFI/list1.xls"
OMB_2023 = "/content/drive/MyDrive/Saksham_MEFI/list1_2023.xlsx"
MEFI_FILE = "/content/drive/MyDrive/Saksham_MEFI/MEFI+2022+dataset+for+researchers,+1-5-23 (4).xlsx"

# ============================================================================
# HELPER FUNCTIONS
# ============================================================================

def z5(x):
    """Pad codes to 5 digits"""
    if pd.isna(x): return pd.NA
    s = str(x).strip()
    if s == "" or s.lower() == "nan": return pd.NA
    if s.replace(".","",1).isdigit():
        s = str(int(float(s)))
    return s.zfill(5)

def search_code_in_df(df, code, name="DataFrame"):
    """Search for a code in all columns of a dataframe"""
    results = {}
    code_padded = z5(code)

    for col in df.columns:
        if df[col].dtype == 'object' or 'int' in str(df[col].dtype) or 'float' in str(df[col].dtype):
            # Convert to string and pad
            df_col_padded = df[col].apply(z5)
            matches = df[df_col_padded == code_padded]

            if len(matches) > 0:
                results[col] = {
                    'count': len(matches),
                    'sample_rows': matches.head(3).to_dict('records')
                }

    return results

# ============================================================================
# LOAD DATA FILES
# ============================================================================

print("\n" + "=" * 80)
print("STEP 1: LOADING OMB LIST 1 FILES")
print("=" * 80)

# Load OMB List 1 (Old)
print("\nLoading old OMB List 1...")
try:
    omb_old = pd.read_excel(OMB_OLD, sheet_name="List 1", dtype=str, header=2)
    omb_old.columns = [str(c).strip() for c in omb_old.columns]
    print(f"  ✓ Loaded: {omb_old.shape[0]:,} rows, {omb_old.shape[1]} columns")
    print(f"  Columns: {omb_old.columns.tolist()[:10]}")

    # Standardize code columns
    for col in omb_old.columns:
        if 'code' in col.lower():
            omb_old[col] = omb_old[col].apply(z5)
except Exception as e:
    print(f"  ✗ Error loading old OMB: {e}")
    omb_old = None

# Load OMB List 1 (2023)
print("\nLoading 2023 OMB List 1...")
try:
    omb_2023 = pd.read_excel(OMB_2023, sheet_name="List 1", dtype=str, header=2)
    omb_2023.columns = [str(c).strip() for c in omb_2023.columns]
    print(f"  ✓ Loaded: {omb_2023.shape[0]:,} rows, {omb_2023.shape[1]} columns")
    print(f"  Columns: {omb_2023.columns.tolist()[:10]}")

    # Standardize code columns
    for col in omb_2023.columns:
        if 'code' in col.lower():
            omb_2023[col] = omb_2023[col].apply(z5)
except Exception as e:
    print(f"  ✗ Error loading 2023 OMB: {e}")
    omb_2023 = None

# Load MEFI
print("\nLoading MEFI file...")
try:
    mefi = pd.read_excel(MEFI_FILE, sheet_name="scores (panel)", dtype=str, header=0)
    mefi.columns = [str(c).strip() for c in mefi.columns]
    print(f"  ✓ Loaded: {mefi.shape[0]:,} rows, {mefi.shape[1]} columns")

    # Standardize CBSA Code
    if 'CBSA Code' in mefi.columns:
        mefi['CBSA Code'] = mefi['CBSA Code'].apply(z5)

    # Get 2017 subset
    if 'year' in mefi.columns:
        mefi_2017 = mefi[mefi['year'].astype(str) == '2017'].copy()
        print(f"  MEFI 2017 entries: {len(mefi_2017):,}")
except Exception as e:
    print(f"  ✗ Error loading MEFI: {e}")
    mefi = None
    mefi_2017 = None

# ============================================================================
# INVESTIGATE EACH MYSTERY CODE
# ============================================================================

print("\n\n" + "=" * 80)
print("STEP 2: INVESTIGATING EACH MYSTERY CODE")
print("=" * 80)

results = {}

for code, description in MYSTERY_CODES.items():
    print(f"\n{'=' * 80}")
    print(f"CODE: {code}")
    print(f"Context: {description}")
    print("=" * 80)

    code_results = {
        'code': code,
        'description': description,
        'omb_old': None,
        'omb_2023': None,
        'mefi': None,
        'recommendation': None
    }

    # Search in old OMB
    if omb_old is not None:
        print("\n🔍 Searching OLD OMB List 1...")
        findings = search_code_in_df(omb_old, code, "Old OMB")
        if findings:
            code_results['omb_old'] = findings
            for col, data in findings.items():
                print(f"  ✓ Found in column '{col}': {data['count']} matches")
                for i, row in enumerate(data['sample_rows'][:2], 1):
                    print(f"    Sample {i}:")
                    # Show key columns
                    key_cols = ['CBSA Code', 'CBSA Title', 'Metropolitan Division Code',
                               'Metropolitan/Micropolitan Statistical Area', 'State Name']
                    for k in key_cols:
                        if k in row and pd.notna(row[k]):
                            print(f"      {k}: {row[k]}")
        else:
            print(f"  ✗ Not found in old OMB")

    # Search in 2023 OMB
    if omb_2023 is not None:
        print("\n🔍 Searching 2023 OMB List 1...")
        findings = search_code_in_df(omb_2023, code, "2023 OMB")
        if findings:
            code_results['omb_2023'] = findings
            for col, data in findings.items():
                print(f"  ✓ Found in column '{col}': {data['count']} matches")
                for i, row in enumerate(data['sample_rows'][:2], 1):
                    print(f"    Sample {i}:")
                    # Show key columns
                    key_cols = ['CBSA Code', 'CBSA Title', 'Metropolitan Division Code',
                               'Metropolitan/Micropolitan Statistical Area', 'State Name']
                    for k in key_cols:
                        if k in row and pd.notna(row[k]):
                            print(f"      {k}: {row[k]}")
        else:
            print(f"  ✗ Not found in 2023 OMB")

    # Search in MEFI
    if mefi_2017 is not None:
        print("\n🔍 Searching MEFI 2017...")
        findings = search_code_in_df(mefi_2017, code, "MEFI")
        if findings:
            code_results['mefi'] = findings
            for col, data in findings.items():
                print(f"  ✓ Found in column '{col}': {data['count']} matches")
                for i, row in enumerate(data['sample_rows'][:2], 1):
                    print(f"    Sample {i}:")
                    # Show key columns
                    key_cols = ['CBSA Code', 'Unnamed: 2', 'Overall', 'year']
                    for k in key_cols:
                        if k in row and pd.notna(row[k]):
                            print(f"      {k}: {row[k]}")
        else:
            print(f"  ✗ Not found in MEFI 2017")

    # Generate recommendation
    print("\n💡 RECOMMENDATION:")

    # Determine what this code is
    is_mdiv = False
    parent_cbsa = None
    is_in_mefi = code_results['mefi'] is not None

    # Check if it's a Metropolitan Division
    if code_results['omb_2023']:
        if 'Metropolitan Division Code' in code_results['omb_2023']:
            is_mdiv = True
            # Find parent CBSA
            samples = code_results['omb_2023']['Metropolitan Division Code']['sample_rows']
            if samples:
                parent_cbsa = samples[0].get('CBSA Code')

    if code_results['omb_old'] and not is_mdiv:
        if 'Metropolitan Division Code' in code_results['omb_old']:
            is_mdiv = True
            samples = code_results['omb_old']['Metropolitan Division Code']['sample_rows']
            if samples:
                parent_cbsa = samples[0].get('CBSA Code')

    # Generate recommendation
    if is_mdiv and parent_cbsa:
        rec = f"ADD TO MDIV MAPPING: '{code}' → '{parent_cbsa}' (Metropolitan Division)"
        print(f"  ✅ {rec}")
    elif is_in_mefi:
        rec = f"EXCLUDE: Code '{code}' is in MEFI but should be excluded (historic MSA)"
        print(f"  🚫 {rec}")
    elif code_results['omb_old'] and not code_results['omb_2023']:
        rec = f"LEGACY CODE: '{code}' exists in old OMB but not 2023 - needs mapping"
        print(f"  ⚠️ {rec}")
    else:
        rec = f"INVESTIGATE MANUALLY: Cannot determine classification"
        print(f"  ❓ {rec}")

    code_results['recommendation'] = rec
    results[code] = code_results

# ============================================================================
# SUMMARY OUTPUT
# ============================================================================

print("\n\n" + "=" * 80)
print("INVESTIGATION SUMMARY")
print("=" * 80)

print("\n📋 RECOMMENDATIONS FOR EACH CODE:\n")
for code, data in results.items():
    print(f"  {code}: {data['recommendation']}")

# Generate mapping additions
print("\n\n" + "=" * 80)
print("MAPPING ADDITIONS NEEDED")
print("=" * 80)

mdiv_mappings = {}
exclusions = []

for code, data in results.items():
    if 'ADD TO MDIV MAPPING' in data['recommendation']:
        # Extract parent CBSA from samples
        if data['omb_2023'] and 'Metropolitan Division Code' in data['omb_2023']:
            samples = data['omb_2023']['Metropolitan Division Code']['sample_rows']
            if samples:
                parent = samples[0].get('CBSA Code')
                if parent:
                    mdiv_mappings[code] = parent
        elif data['omb_old'] and 'Metropolitan Division Code' in data['omb_old']:
            samples = data['omb_old']['Metropolitan Division Code']['sample_rows']
            if samples:
                parent = samples[0].get('CBSA Code')
                if parent:
                    mdiv_mappings[code] = parent

    elif 'EXCLUDE' in data['recommendation']:
        exclusions.append(code)

if mdiv_mappings:
    print("\n✅ Metropolitan Division → Parent MSA mappings to add:")
    print("```python")
    for mdiv, parent in mdiv_mappings.items():
        print(f'    "{mdiv}": "{parent}",  # Add to div2cbsa or legacy_to_current')
    print("```")

if exclusions:
    print("\n🚫 Codes to exclude:")
    print("```python")
    print(f"excluded_historic = {{{', '.join([f'\"{c}\"' for c in exclusions])}}}")
    print("```")

# Save results to JSON
output_file = "/content/mystery_code_investigation_results.json"
with open(output_file, 'w') as f:
    json.dump(results, f, indent=2, default=str)

print(f"\n\n💾 Detailed results saved to: {output_file}")

print("\n" + "=" * 80)
print("✨ INVESTIGATION COMPLETE!")
print("=" * 80)
print("\nNext steps:")
print("  1. Review the recommendations above")
print("  2. Run OPTION 2 to extract comprehensive MDiv mappings")
print("  3. Use both outputs to build the patch file")
````

````python
"""
OPTION 2: COMPREHENSIVE MDIV & EXCLUSION EXTRACTION
Purpose: Build complete mapping dictionaries from OMB List 1

This script extracts:
1. ALL Metropolitan Division → Parent MSA mappings
2. ALL Micropolitan areas
3. ALL Puerto Rico & territory metros
4. MSA names for both MDiv and parent (for new columns)
"""

import pandas as pd
import numpy as np
import json

print("=" * 80)
print("COMPREHENSIVE MAPPING EXTRACTION")
print("=" * 80)

# ============================================================================
# CONFIGURATION
# ============================================================================

# File paths
OMB_OLD = "/content/drive/MyDrive/Saksham_MEFI/list1.xls"
OMB_2023 = "/content/drive/MyDrive/Saksham_MEFI/list1_2023.xlsx"

# ============================================================================
# HELPER FUNCTIONS
# ============================================================================

def z5(x):
    """Pad codes to 5 digits"""
    if pd.isna(x): return pd.NA
    s = str(x).strip()
    if s == "" or s.lower() == "nan": return pd.NA
    if s.replace(".","",1).isdigit():
        s = str(int(float(s)))
    return s.zfill(5)

# ============================================================================
# LOAD OMB FILES
# ============================================================================

print("\n" + "=" * 80)
print("STEP 1: LOADING OMB LIST 1 FILES")
print("=" * 80)

# Load both versions
omb_files = {}

for name, path, header_row in [("Old", OMB_OLD, 2), ("2023", OMB_2023, 2)]:
    print(f"\nLoading {name} OMB List 1...")
    try:
        df = pd.read_excel(path, sheet_name="List 1", dtype=str, header=header_row)
        df.columns = [str(c).strip() for c in df.columns]

        # Standardize code columns
        for col in df.columns:
            if 'code' in col.lower() and col not in ['Central/Outlying County']:
                df[col] = df[col].apply(z5)

        omb_files[name] = df
        print(f"  ✓ Loaded: {df.shape[0]:,} rows, {df.shape[1]} columns")
        print(f"  Key columns: {[c for c in df.columns if 'code' in c.lower() or 'title' in c.lower()][:5]}")

    except Exception as e:
        print(f"  ✗ Error: {e}")
        omb_files[name] = None

# ============================================================================
# EXTRACT METROPOLITAN DIVISION MAPPINGS
# ============================================================================

print("\n\n" + "=" * 80)
print("STEP 2: EXTRACTING METROPOLITAN DIVISION MAPPINGS")
print("=" * 80)

all_mdiv_mappings = {}
mdiv_names = {}  # Store MDiv names for new columns
parent_msa_names = {}  # Store parent MSA names

for version, df in omb_files.items():
    if df is None:
        continue

    print(f"\n--- Processing {version} OMB ---")

    # Find relevant columns
    mdiv_col = None
    cbsa_col = None
    cbsa_title_col = None

    for col in df.columns:
        col_lower = col.lower()
        if 'division' in col_lower and 'code' in col_lower:
            mdiv_col = col
        elif 'cbsa' in col_lower and 'code' in col_lower and 'division' not in col_lower:
            cbsa_col = col
        elif 'cbsa' in col_lower and 'title' in col_lower:
            cbsa_title_col = col

    print(f"  Metropolitan Division Column: {mdiv_col}")
    print(f"  CBSA Code Column: {cbsa_col}")
    print(f"  CBSA Title Column: {cbsa_title_col}")

    if not mdiv_col or not cbsa_col:
        print(f"  ⚠️ Missing required columns, skipping...")
        continue

    # Extract mappings
    mdiv_data = df[[mdiv_col, cbsa_col]].dropna(subset=[mdiv_col])

    # Also get titles if available
    if cbsa_title_col:
        mdiv_data = df[[mdiv_col, cbsa_col, cbsa_title_col]].dropna(subset=[mdiv_col])

    mdiv_data = mdiv_data.drop_duplicates(subset=[mdiv_col])

    print(f"  Found {len(mdiv_data)} Metropolitan Division mappings")

    # Build mapping dictionary
    for _, row in mdiv_data.iterrows():
        mdiv_code = row[mdiv_col]
        parent_code = row[cbsa_col]

        if pd.notna(mdiv_code) and pd.notna(parent_code):
            all_mdiv_mappings[mdiv_code] = parent_code

            # Store title if available
            if cbsa_title_col and cbsa_title_col in row.index:
                parent_title = row[cbsa_title_col]
                if pd.notna(parent_title):
                    parent_msa_names[parent_code] = parent_title

    # Show sample
    print(f"\n  Sample mappings from {version}:")
    for i, (mdiv, parent) in enumerate(list(all_mdiv_mappings.items())[:5], 1):
        parent_name = parent_msa_names.get(parent, "N/A")
        print(f"    {i}. {mdiv} → {parent} ({parent_name})")

print(f"\n✅ TOTAL METROPOLITAN DIVISION MAPPINGS: {len(all_mdiv_mappings)}")

# ============================================================================
# EXTRACT MICROPOLITAN AREAS
# ============================================================================

print("\n\n" + "=" * 80)
print("STEP 3: EXTRACTING MICROPOLITAN AREAS")
print("=" * 80)

all_micros = set()
micro_names = {}

for version, df in omb_files.items():
    if df is None:
        continue

    print(f"\n--- Processing {version} OMB ---")

    # Find columns
    cbsa_col = None
    mm_type_col = None
    title_col = None

    for col in df.columns:
        col_lower = col.lower()
        if 'cbsa' in col_lower and 'code' in col_lower and 'division' not in col_lower:
            cbsa_col = col
        elif 'metropolitan' in col_lower and 'micropolitan' in col_lower:
            mm_type_col = col
        elif 'cbsa' in col_lower and 'title' in col_lower:
            title_col = col

    print(f"  CBSA Code Column: {cbsa_col}")
    print(f"  Metro/Micro Type Column: {mm_type_col}")
    print(f"  Title Column: {title_col}")

    if not cbsa_col or not mm_type_col:
        print(f"  ⚠️ Missing required columns, skipping...")
        continue

    # Filter for Micropolitan
    micros = df[df[mm_type_col].astype(str).str.contains('Micropolitan', case=False, na=False)]
    micros = micros[[cbsa_col, title_col]].drop_duplicates(subset=[cbsa_col]) if title_col else micros[[cbsa_col]].drop_duplicates(subset=[cbsa_col])

    print(f"  Found {len(micros)} Micropolitan areas")

    for _, row in micros.iterrows():
        code = row[cbsa_col]
        if pd.notna(code):
            all_micros.add(code)
            if title_col and title_col in row.index and pd.notna(row[title_col]):
                micro_names[code] = row[title_col]

    # Show sample
    print(f"\n  Sample micros from {version}:")
    for i, code in enumerate(list(all_micros)[:5], 1):
        name = micro_names.get(code, "N/A")
        print(f"    {i}. {code}: {name}")

print(f"\n✅ TOTAL MICROPOLITAN AREAS: {len(all_micros)}")

# ============================================================================
# EXTRACT PUERTO RICO & TERRITORIES
# ============================================================================

print("\n\n" + "=" * 80)
print("STEP 4: EXTRACTING PUERTO RICO & US TERRITORIES")
print("=" * 80)

all_pr_territory = set()
pr_territory_names = {}

territories = ['Puerto Rico', 'Virgin Islands', 'Guam', 'Northern Mariana Islands', 'American Samoa']

for version, df in omb_files.items():
    if df is None:
        continue

    print(f"\n--- Processing {version} OMB ---")

    # Find columns
    cbsa_col = None
    state_col = None
    title_col = None

    for col in df.columns:
        col_lower = col.lower()
        if 'cbsa' in col_lower and 'code' in col_lower and 'division' not in col_lower:
            cbsa_col = col
        elif 'state' in col_lower and 'name' in col_lower:
            state_col = col
        elif 'cbsa' in col_lower and 'title' in col_lower:
            title_col = col

    print(f"  CBSA Code Column: {cbsa_col}")
    print(f"  State Name Column: {state_col}")
    print(f"  Title Column: {title_col}")

    if not cbsa_col or not state_col:
        print(f"  ⚠️ Missing required columns, skipping...")
        continue

    # Filter for territories
    for territory in territories:
        territory_rows = df[df[state_col].astype(str).str.contains(territory, case=False, na=False)]
        if len(territory_rows) > 0:
            print(f"  Found {len(territory_rows)} rows for {territory}")

            for _, row in territory_rows.iterrows():
                code = row[cbsa_col]
                if pd.notna(code):
                    all_pr_territory.add(code)
                    if title_col and title_col in row.index and pd.notna(row[title_col]):
                        pr_territory_names[code] = f"{row[title_col]} ({territory})"

print(f"\n✅ TOTAL PR/TERRITORY METROS: {len(all_pr_territory)}")
for code in sorted(all_pr_territory):
    name = pr_territory_names.get(code, "N/A")
    print(f"  {code}: {name}")

# ============================================================================
# EXTRACT CBSA NAMES FOR ALL CODES
# ============================================================================

print("\n\n" + "=" * 80)
print("STEP 5: EXTRACTING ALL CBSA NAMES")
print("=" * 80)

all_cbsa_names = {}

# Use 2023 as primary source
if "2023" in omb_files and omb_files["2023"] is not None:
    df = omb_files["2023"]

    cbsa_col = None
    title_col = None

    for col in df.columns:
        col_lower = col.lower()
        if 'cbsa' in col_lower and 'code' in col_lower and 'division' not in col_lower:
            cbsa_col = col
        elif 'cbsa' in col_lower and 'title' in col_lower:
            title_col = col

    if cbsa_col and title_col:
        cbsa_data = df[[cbsa_col, title_col]].drop_duplicates(subset=[cbsa_col])

        for _, row in cbsa_data.iterrows():
            code = row[cbsa_col]
            title = row[title_col]
            if pd.notna(code) and pd.notna(title):
                all_cbsa_names[code] = title

        print(f"  Extracted {len(all_cbsa_names)} CBSA names from 2023 OMB")

# ============================================================================
# BUILD OUTPUT DICTIONARIES
# ============================================================================

print("\n\n" + "=" * 80)
print("STEP 6: BUILDING OUTPUT DICTIONARIES")
print("=" * 80)

output = {
    "div2cbsa": all_mdiv_mappings,
    "micropolitan_areas": sorted(list(all_micros)),
    "puerto_rico_territories": sorted(list(all_pr_territory)),
    "cbsa_names": all_cbsa_names,
    "mdiv_names": mdiv_names,
    "parent_msa_names": parent_msa_names,
    "micropolitan_names": micro_names,
    "territory_names": pr_territory_names,
    "metadata": {
        "total_mdiv_mappings": len(all_mdiv_mappings),
        "total_micros": len(all_micros),
        "total_territories": len(all_pr_territory),
        "total_cbsa_names": len(all_cbsa_names),
        "source": "OMB List 1 (Old + 2023)",
    }
}

# Print summaries
print("\n📊 EXTRACTION SUMMARY:")
print(f"  Metropolitan Divisions: {len(all_mdiv_mappings)}")
print(f"  Micropolitan Areas: {len(all_micros)}")
print(f"  Puerto Rico/Territories: {len(all_pr_territory)}")
print(f"  CBSA Names: {len(all_cbsa_names)}")

# ============================================================================
# GENERATE PYTHON CODE SNIPPETS
# ============================================================================

print("\n\n" + "=" * 80)
print("STEP 7: GENERATING PYTHON CODE FOR YOUR SCRIPT")
print("=" * 80)

print("\n# ========== Metropolitan Division → Parent MSA Mapping ==========")
print("div2cbsa = {")
for mdiv, parent in sorted(all_mdiv_mappings.items()):
    parent_name = parent_msa_names.get(parent, "")
    if parent_name:
        print(f'    "{mdiv}": "{parent}",  # → {parent_name}')
    else:
        print(f'    "{mdiv}": "{parent}",')
print("}")

print("\n\n# ========== Micropolitan Areas (Exclude) ==========")
print("micropolitan_areas = {")
for code in sorted(all_micros):
    name = micro_names.get(code, "")
    if name:
        print(f'    "{code}",  # {name}')
    else:
        print(f'    "{code}",')
print("}")

print("\n\n# ========== Puerto Rico & Territories (Exclude) ==========")
print("puerto_rico_territories = {")
for code in sorted(all_pr_territory):
    name = pr_territory_names.get(code, "")
    if name:
        print(f'    "{code}",  # {name}')
    else:
        print(f'    "{code}",')
print("}")

# ============================================================================
# SAVE TO FILES
# ============================================================================

print("\n\n" + "=" * 80)
print("STEP 8: SAVING OUTPUT FILES")
print("=" * 80)

# Save as JSON
json_file = "/content/comprehensive_mappings.json"
with open(json_file, 'w') as f:
    json.dump(output, f, indent=2, default=str)
print(f"  ✓ Saved JSON: {json_file}")

# Save as Python file (importable)
py_file = "/content/comprehensive_mappings.py"
with open(py_file, 'w') as f:
    f.write('"""\nComprehensive CBSA Mappings\n')
    f.write('Extracted from OMB List 1 (Old + 2023)\n')
    f.write(f'Total MDivs: {len(all_mdiv_mappings)}\n')
    f.write(f'Total Micros: {len(all_micros)}\n')
    f.write(f'Total Territories: {len(all_pr_territory)}\n')
    f.write('"""\n\n')

    f.write("# Metropolitan Division → Parent MSA\n")
    f.write("DIV2CBSA = {\n")
    for mdiv, parent in sorted(all_mdiv_mappings.items()):
        f.write(f'    "{mdiv}": "{parent}",\n')
    f.write("}\n\n")

    f.write("# Micropolitan Areas\n")
    f.write("MICROPOLITAN = {\n")
    for code in sorted(all_micros):
        f.write(f'    "{code}",\n')
    f.write("}\n\n")

    f.write("# Puerto Rico & Territories\n")
    f.write("TERRITORIES = {\n")
    for code in sorted(all_pr_territory):
        f.write(f'    "{code}",\n')
    f.write("}\n\n")

    f.write("# CBSA Names (2023)\n")
    f.write("CBSA_NAMES = {\n")
    for code, name in sorted(all_cbsa_names.items()):
        f.write(f'    "{code}": "{name}",\n')
    f.write("}\n")

print(f"  ✓ Saved Python: {py_file}")

# ============================================================================
# COMPARISON WITH CURRENT CODE
# ============================================================================

print("\n\n" + "=" * 80)
print("STEP 9: COMPARISON WITH YOUR CURRENT CODE")
print("=" * 80)

# Your current hardcoded values
current_mdivs = {
    "14454": "14460", "15764": "14460", "40484": "14460",
    "19124": "19100", "23104": "19100",
    "47664": "19820",
    "35004": "35620", "35084": "35620", "35614": "35620",
    "15804": "37980", "33874": "37980",
    "17200": "30150", "19430": "19380",
}

current_micros = {"10100", "25740", "33500", "35820"}
current_pr = {"10380", "41980"}

print(f"\n📊 COMPARISON:")
print(f"\n  Metropolitan Divisions:")
print(f"    Your current code: {len(current_mdivs)} mappings")
print(f"    Comprehensive extraction: {len(all_mdiv_mappings)} mappings")
print(f"    NEW mappings to add: {len(all_mdiv_mappings) - len(current_mdivs)}")

print(f"\n  Micropolitan Areas:")
print(f"    Your current code: {len(current_micros)} areas")
print(f"    Comprehensive extraction: {len(all_micros)} areas")
print(f"    NEW areas to add: {len(all_micros) - len(current_micros)}")

print(f"\n  Puerto Rico/Territories:")
print(f"    Your current code: {len(current_pr)} metros")
print(f"    Comprehensive extraction: {len(all_pr_territory)} metros")
print(f"    NEW metros to add: {len(all_pr_territory) - len(current_pr)}")

# Show what's missing from current code
new_mdivs = set(all_mdiv_mappings.keys()) - set(current_mdivs.keys())
if new_mdivs:
    print(f"\n  🆕 NEW MDiv mappings found (sample of first 10):")
    for code in sorted(list(new_mdivs))[:10]:
        parent = all_mdiv_mappings[code]
        parent_name = parent_msa_names.get(parent, "N/A")
        print(f"    {code} → {parent} ({parent_name})")

print("\n" + "=" * 80)
print("✨ COMPREHENSIVE EXTRACTION COMPLETE!")
print("=" * 80)
print("\nNext steps:")
print("  1. Review the generated Python code above")
print("  2. Review saved files:")
print(f"     - {json_file}")
print(f"     - {py_file}")
print("  3. Combine with OPTION 1 results to build patch file")
````

````python
"""
BRFSS-MEFI DATA REFINEMENT PATCH FILE
Generated: Based on Option 1 & 2 investigation results

This patch updates your existing code with:
1. Comprehensive Metropolitan Division → MSA mappings (13 → 42)
2. Complete Micropolitan areas list (4 → 603)
3. Complete Puerto Rico/Territories list (2 → 14)
4. New columns for MDiv/MSA names
5. Poughkeepsie exclusion

EXPECTED IMPACT:
- Match rate: 92% → 97%+
- Unmatched rows: 14,000 → ~3,800
- Fixes 10,000+ rows with DC MDivs and micropolitan codes
"""

import pandas as pd
import numpy as np
import json

# ============================================================================
# SECTION 1: COMPREHENSIVE MAPPING DICTIONARIES
# Replace your existing dictionaries with these
# ============================================================================

# ========== Metropolitan Division → Parent MSA Mapping ==========
# OLD: You had 13 mappings
# NEW: 42 mappings (from OMB List 1)
DIV2CBSA = {
    "11244": "31080",  # Los Angeles MDivs
    "31084": "31080",

    "11694": "47900",  # Washington DC MDivs
    "23224": "47900",
    "43524": "47900",  # Silver Spring-Frederick-Rockville, MD
    "47764": "47900",
    "47894": "47900",  # Washington-Arlington-Alexandria, DC-VA-MD-WV ← FIXES 8,500 ROWS!

    "12054": "12060",  # Atlanta MDivs
    "31924": "12060",

    "14454": "14460",  # Boston MDivs
    "15764": "14460",
    "40484": "14460",

    "15804": "37980",  # Philadelphia MDivs
    "33874": "37980",
    "37964": "37980",
    "48864": "37980",

    "16974": "16980",  # Chicago MDivs
    "16984": "16980",
    "20994": "16980",
    "23844": "16980",
    "29404": "16980",
    "29414": "16980",

    "19124": "19100",  # Dallas MDivs
    "23104": "19100",

    "19804": "19820",  # Detroit MDivs
    "47664": "19820",

    "20524": "35620",  # New York MDivs
    "29484": "35620",
    "35004": "35620",
    "35084": "35620",
    "35614": "35620",

    "21794": "42660",  # Seattle MDivs
    "42644": "42660",
    "45104": "42660",

    "22744": "33100",  # Miami MDivs
    "33124": "33100",
    "48424": "33100",

    "36084": "41860",  # San Francisco MDivs
    "41884": "41860",
    "42034": "41860",

    "41304": "45300",  # Tampa MDivs
    "45294": "45300",
}

# ========== Legacy Code Renames (Keep your existing ones) ==========
LEGACY_RENAMES = {
    "17200": "30150",  # Claremont-Lebanon → Lebanon-Claremont
    "19430": "19380",  # Dayton-Springfield → Dayton
}

# ========== Micropolitan Areas (EXCLUDE) ==========
# OLD: You had 4 hardcoded
# NEW: 603 comprehensive (all micros from OMB)
MICROPOLITAN_AREAS = {
    "10100", "10140", "10220", "10260", "10300", "10460", "10480", "10620",
    "10660", "10700", "10760", "10820", "10860", "10940", "10980", "11060",
    "11140", "11220", "11360", "11380", "11420", "11580", "11620", "11660",
    "11680", "11740", "11780", "11820", "11860", "11900", "11940", "11980",
    "12140", "12180", "12300", "12380", "12460", "12520", "12660", "12680",
    "12740", "12780", "12820", "12860", "12900", "13060", "13100", "13180",
    "13260", "13300", "13340", "13420", "13500", "13540", "13620", "13660",
    "13700", "13720", "13860", "13940", "14100", "14140", "14180", "14220",
    "14300", "14340", "14380", "14420", "14580", "14620", "14660", "14700",
    "14710", "14720", "14780", "14820", "14940", "15020", "15060", "15100",
    "15220", "15340", "15420", "15460", "15580", "15620", "15660", "15700",
    "15740", "15780", "15820", "15860", "15900", "16060", "16100", "16140",
    "16260", "16340", "16380", "16460", "16500", "16660", "17060", "17200",
    "17220", "17260", "17340", "17380", "17500", "17540", "17580", "17620",
    "17640", "17650", "17700", "17740", "18060", "18100", "18180", "18220",
    "18260", "18300", "18340", "18380", "18420", "18460", "18500", "18620",
    "18660", "18740", "18780", "18820", "18860", "18900", "18980", "19000",
    "19060", "19180", "19220", "19260", "19420", "19540", "19580", "19620",
    "19700", "19760", "19810", "19860", "19940", "19980", "20060", "20140",
    "20180", "20300", "20340", "20380", "20420", "20460", "20540", "20580",
    "20660", "20700", "20780", "20820", "20900", "20980", "21020", "21120",
    "21180", "21220", "21260", "21380", "21420", "21460", "21540", "21580",
    "21640", "21700", "21740", "21840", "21860", "21900", "21980", "22060",
    "22100", "22190", "22260", "22280", "22300", "22340", "22580", "22620",
    "22700", "22780", "22800", "22820", "22840", "22860", "23140", "23180",
    "23190", "23240", "23300", "23340", "23380", "23500", "23620", "23660",
    "23680", "23700", "23780", "23820", "23860", "23940", "23980", "24060",
    "24100", "24180", "24330", "24380", "24460", "24600", "24620", "24640",
    "24700", "24740", "24820", "24900", "24940", "24980", "25100", "25200",
    "25300", "25460", "25580", "25700", "25720", "25740", "25760", "25770",
    "25775", "25780", "25820", "25840", "25850", "25880", "25900", "26020",
    "26090", "26220", "26340", "26460", "26500", "26540", "26660", "26700",
    "26740", "26780", "26860", "26940", "26960", "27020", "27160", "27220",
    "27300", "27380", "27420", "27460", "27540", "27580", "27600", "27700",
    "27920", "27940", "28060", "28180", "28260", "28300", "28340", "28380",
    "28500", "28540", "28580", "28620", "28680", "28780", "28820", "28860",
    "28900", "29060", "29260", "29300", "29380", "29470", "29500", "29660",
    "29720", "29780", "29860", "29900", "29980", "30060", "30150", "30180",
    "30220", "30260", "30280", "30380", "30420", "30580", "30660", "30820",
    "30880", "30900", "30940", "31060", "31220", "31260", "31300", "31380",
    "31500", "31580", "31620", "31660", "31680", "31820", "31930", "31940",
    "31980", "32000", "32020", "32060", "32100", "32140", "32180", "32220",
    "32260", "32280", "32300", "32340", "32380", "32390", "32460", "32500",
    "32540", "32620", "32660", "32700", "32740", "32860", "32940", "32980",
    "33020", "33060", "33180", "33300", "33380", "33420", "33500", "33580",
    "33620", "33900", "33910", "33940", "33980", "34020", "34140", "34180",
    "34220", "34260", "34300", "34340", "34380", "34420", "34460", "34500",
    "34540", "34660", "34680", "34700", "34780", "34860", "34880", "35020",
    "35060", "35100", "35140", "35220", "35260", "35340", "35420", "35440",
    "35460", "35500", "35580", "35700", "35740", "35820", "35860", "35900",
    "35940", "36020", "36180", "36300", "36340", "36380", "36460", "36580",
    "36620", "36660", "36700", "36820", "36830", "36837", "36840", "36860",
    "36900", "36940", "37020", "37060", "37080", "37120", "37140", "37220",
    "37260", "37300", "37420", "37500", "37540", "37580", "37660", "37740",
    "37780", "37800", "37940", "37950", "38100", "38180", "38210", "38220",
    "38240", "38260", "38380", "38420", "38460", "38500", "38580", "38620",
    "38700", "38740", "38780", "38820", "38840", "38920", "39020", "39040",
    "39060", "39220", "39260", "39420", "39480", "39500", "39700", "39780",
    "39860", "39940", "39960", "39980", "40080", "40090", "40100", "40180",
    "40260", "40300", "40460", "40540", "40620", "40700", "40740", "40760",
    "40770", "40780", "40820", "40860", "40940", "41220", "41260", "41400",
    "41460", "41760", "41780", "41820", "42180", "42300", "42380", "42420",
    "42460", "42580", "42620", "42740", "42780", "42820", "42860", "42900",
    "42940", "42980", "43020", "43060", "43140", "43180", "43220", "43260",
    "43320", "43380", "43460", "43500", "43660", "43700", "43740", "43760",
    "43890", "43940", "43980", "44020", "44260", "44340", "44460", "44500",
    "44540", "44580", "44620", "44660", "44740", "44780", "44860", "44900",
    "44920", "44980", "45000", "45020", "45140", "45180", "45340", "45380",
    "45520", "45580", "45620", "45660", "45700", "45740", "45860", "45880",
    "45900", "45980", "46020", "46100", "46180", "46300", "46380", "46460",
    "46500", "46580", "46620", "46740", "46780", "46820", "46860", "46900",
    "46980", "47080", "47180", "47240", "47340", "47420", "47540", "47620",
    "47660", "47700", "47780", "47820", "47920", "47980", "48020", "48100",
    "48180", "48200", "48220", "48460", "48580", "48780", "48820", "48940",
    "48980", "49010", "49080", "49100", "49220", "49260", "49300", "49380",
    "49460", "49780", "49820"
}

# ========== Puerto Rico & Territories (EXCLUDE) ==========
# OLD: You had 2 hardcoded
# NEW: 14 comprehensive (all PR metros from OMB)
PUERTO_RICO_TERRITORIES = {
    "10260",  # Adjuntas, PR
    "10380",  # Aguadilla, PR
    "11640",  # Arecibo, PR
    "17620",  # Coamo, PR
    "17640",  # Coco, PR
    "25020",  # Guayama, PR
    "27580",  # Jayuya, PR
    "29720",  # Lares, PR
    "32420",  # Mayagüez, PR
    "38660",  # Ponce, PR
    "41900",  # San Germán, PR
    "41980",  # San Juan-Bayamón-Caguas, PR
    "42180",  # Santa Isabel, PR
    "46580",  # Utuado, PR
}

# ========== Historic MSAs to EXCLUDE ==========
# NEW: Based on investigation, Poughkeepsie should be excluded
EXCLUDED_HISTORIC_MSAS = {
    "39100",  # Poughkeepsie-Newburgh-Middletown, NY (appears in 2022, not in MEFI properly)
}

# ========== Mystery Codes Still Need Investigation ==========
# These codes appeared in your data but weren't found in OMB or MEFI
# TODO: Look at actual BRFSS data to see what metro names these have
# MYSTERY_CODES_TODO = {
#     "35154",  # 2,235 rows in 2022 - unknown
#     "30100",  # 2,066 rows in 2022 - possibly Lebanon area related
# }


# ============================================================================
# SECTION 2: UPDATED HELPER FUNCTIONS
# Keep your existing z5() and norm_name(), add these new ones
# ============================================================================

def z5(x):
    """Pad CBSA codes to 5 digits with leading zeros"""
    if pd.isna(x): return pd.NA
    s = str(x).strip()
    if s == "" or s.lower() == "nan": return pd.NA
    if s.replace(".","",1).isdigit():
        s = str(int(float(s)))
    return s.zfill(5)

def norm_name(s):
    """Normalize metro area names for matching"""
    if pd.isna(s): return ""
    s = str(s).lower()
    for ch in ["–","—","(",")",".",","]:
        s = s.replace(ch, " ")
    s = s.replace("-", " ")
    return " ".join(s.split())

def add_scope_flags(df, code_col="CBSA_CODE_CANON"):
    """
    Add comprehensive exclusion flags
    UPDATED: Now uses comprehensive sets from OMB extraction
    """
    df["is_micro"] = df[code_col].isin(MICROPOLITAN_AREAS).astype(int)
    df["is_pr"] = df[code_col].isin(PUERTO_RICO_TERRITORIES).astype(int)
    df["is_excluded_historic"] = df[code_col].isin(EXCLUDED_HISTORIC_MSAS).astype(int)

    def _note(code):
        if code in MICROPOLITAN_AREAS:
            return "Micropolitan area — MEFI not available"
        if code in PUERTO_RICO_TERRITORIES:
            return "Puerto Rico/Territory — MEFI typically excluded"
        if code in EXCLUDED_HISTORIC_MSAS:
            return "Historic MSA in MEFI — excluded per protocol"
        if code in MYSTERY_CODES_TODO:
            return "Under investigation — code not in OMB"
        return ""

    df["scope_note"] = df[code_col].map(_note)

    # Master exclusion flag
    df["exclude_from_analysis"] = (
        (df["is_micro"] == 1) |
        (df["is_pr"] == 1) |
        (df["is_excluded_historic"] == 1)
    ).astype(int)

    return df


# ============================================================================
# SECTION 3: UPDATED prep_brfss FUNCTION
# Replace your existing prep_brfss function with this
# ============================================================================

def prep_brfss(df):
    """
    Prepare BRFSS data with comprehensive CBSA code standardization

    UPDATED: Now includes:
    - Comprehensive MDiv rollup (42 mappings)
    - MDiv name retention
    - MSA name lookup
    - Comprehensive exclusion flags
    """
    df = df.copy()
    assert "CBSA_CODE" in df.columns, "CBSA_CODE missing in BRFSS file."

    # Step 1: Standardize original code
    df["CBSA_CODE_ORIG"] = df["CBSA_CODE"].apply(z5)

    # Step 2: Roll up Metropolitan Divisions to parent MSAs
    df["CBSA_CODE_ROLLED"] = df["CBSA_CODE_ORIG"].map(DIV2CBSA).fillna(df["CBSA_CODE_ORIG"])

    # Step 3: Apply legacy renames (e.g., Dayton, Lebanon area)
    df["CBSA_CODE_CANON"] = df["CBSA_CODE_ROLLED"].replace(LEGACY_RENAMES)

    # Step 4: Flag if original was a Metropolitan Division
    df["is_mdiv"] = df["CBSA_CODE_ORIG"].isin(DIV2CBSA.keys()).astype(int)

    # Step 5: Add comprehensive exclusion flags
    df = add_scope_flags(df, "CBSA_CODE_CANON")

    # Step 6: Store original metro name from BRFSS
    if "MMSANAME" in df.columns:
        df["BRFSS_Metro_Name_Orig"] = df["MMSANAME"]

    # Step 7: Drop original CBSA_CODE to avoid confusion
    df = df.drop(columns=["CBSA_CODE"], errors="ignore")

    return df


# ============================================================================
# SECTION 4: ADD MSA/MDIV NAMES FROM OMB
# NEW: This adds the name columns you requested
# ============================================================================

def add_msa_names_from_omb(brfss_df, omb_2023_path):
    """
    Add MSA names from OMB List 1 (2023) to BRFSS data
    Creates columns: MDiv_Name, MSA_Name_Current

    Args:
        brfss_df: BRFSS DataFrame (after prep_brfss)
        omb_2023_path: Path to list1_2023.xlsx file

    Returns:
        DataFrame with added name columns
    """
    # Load OMB 2023
    omb = pd.read_excel(omb_2023_path, sheet_name="List 1", dtype=str, header=2)
    omb.columns = [str(c).strip() for c in omb.columns]

    # Standardize codes
    if "CBSA Code" in omb.columns:
        omb["CBSA Code"] = omb["CBSA Code"].apply(z5)
    if "Metropolitan Division Code" in omb.columns:
        omb["Metropolitan Division Code"] = omb["Metropolitan Division Code"].apply(z5)

    # Create MDiv name lookup
    mdiv_names = {}
    if "Metropolitan Division Code" in omb.columns and "Metropolitan Division Title" in omb.columns:
        mdiv_data = omb[["Metropolitan Division Code", "Metropolitan Division Title"]].dropna()
        mdiv_data = mdiv_data.drop_duplicates(subset=["Metropolitan Division Code"])
        mdiv_names = dict(mdiv_data.values)

    # Create CBSA name lookup (for parent MSAs)
    cbsa_names = {}
    if "CBSA Code" in omb.columns and "CBSA Title" in omb.columns:
        cbsa_data = omb[["CBSA Code", "CBSA Title"]].dropna()
        cbsa_data = cbsa_data.drop_duplicates(subset=["CBSA Code"])
        cbsa_names = dict(cbsa_data.values)

    # Add MDiv name if original code was a MDiv
    brfss_df["MDiv_Name"] = brfss_df.apply(
        lambda row: mdiv_names.get(row["CBSA_CODE_ORIG"]) if row["is_mdiv"] == 1 else pd.NA,
        axis=1
    )

    # Add current MSA name (from canonical code)
    brfss_df["MSA_Name_Current"] = brfss_df["CBSA_CODE_CANON"].map(cbsa_names)

    return brfss_df


# ============================================================================
# SECTION 5: UPDATED MERGE FUNCTION
# Replace your existing merge_brfss_mefi function with this
# ============================================================================

def merge_brfss_mefi(brfss_df, mefi_df, year_label="2017"):
    """
    Merge BRFSS data with MEFI scores and add comprehensive join status flags

    UPDATED: Now handles exclusions and mystery codes properly
    """
    out = brfss_df.merge(mefi_df, left_on="CBSA_CODE_CANON", right_on="CBSA_CODE", how="left")

    mefi_core = [f"MEFI{year_label}_overall", f"MEFI{year_label}_area1",
                 f"MEFI{year_label}_area2", f"MEFI{year_label}_area3"]
    mefi_core_existing = [c for c in mefi_core if c in out.columns]

    out["mefi_joined"] = out[mefi_core_existing].notna().any(axis=1).astype(int)

    # Updated join_status to handle exclusions
    out["join_status"] = np.where(
        out["mefi_joined"].eq(1),
        "joined",
        np.where(
            out["exclude_from_analysis"].eq(1),
            "excluded",
            "unmatched"
        )
    )

    # Updated mefi_join_note with more detail
    out["mefi_join_note"] = np.where(
        out["mefi_joined"].eq(1),
        "MEFI attached",
        np.where(
            out["scope_note"].ne(""),
            out["scope_note"],
            "No MEFI match — check mapping"
        )
    )

    return out


# ============================================================================
# SECTION 6: MAIN EXECUTION EXAMPLE
# Replace your main execution block with this pattern
# ============================================================================

if __name__ == "__main__":

    print("=" * 80)
    print("APPLYING COMPREHENSIVE BRFSS-MEFI REFINEMENT PATCH")
    print("=" * 80)

    # ========= File paths (UPDATE THESE) =========
    SMART2017 = "/content/drive/MyDrive/Saksham_MEFI/SMART_2017_individual.csv"
    SMART2022 = "/content/drive/MyDrive/Saksham_MEFI/SMART_2022_individual.csv"
    MEFI_XLSX = "/content/drive/MyDrive/Saksham_MEFI/MEFI+2022+dataset+for+researchers,+1-5-23 (4).xlsx"
    OMB_2023 = "/content/drive/MyDrive/Saksham_MEFI/list1_2023.xlsx"

    # ========= Load BRFSS data =========
    print("\nStep 1: Loading BRFSS data...")
    B17 = pd.read_csv(SMART2017, dtype=str, low_memory=False)
    B22 = pd.read_csv(SMART2022, dtype=str, low_memory=False)
    print(f"  2017: {len(B17):,} rows")
    print(f"  2022: {len(B22):,} rows")

    # ========= Apply comprehensive prep =========
    print("\nStep 2: Applying comprehensive CBSA standardization...")
    B17 = prep_brfss(B17)
    B22 = prep_brfss(B22)
    print("  ✓ Applied MDiv rollup (42 mappings)")
    print("  ✓ Applied legacy renames")
    print("  ✓ Applied exclusion flags (603 micros, 14 PR/territories)")

    # ========= Add MSA/MDiv names =========
    print("\nStep 3: Adding MSA/MDiv names from OMB...")
    B17 = add_msa_names_from_omb(B17, OMB_2023)
    B22 = add_msa_names_from_omb(B22, OMB_2023)
    print("  ✓ Added MDiv_Name column")
    print("  ✓ Added MSA_Name_Current column")

    # ========= Load and prep MEFI data =========
    print("\nStep 4: Loading MEFI data...")
    scores = pd.read_excel(MEFI_XLSX, sheet_name="scores (panel)", header=0)
    scores = scores.rename(columns={
        "CBSA Code":"CBSA_CODE",
        "Area 1: Govern-ment Spending":"Area1_GovtSpending",
        "Area 2: Taxes":"Area2_Taxes",
        "Area 3: Labor Market Regulation":"Area3_Labor",
    })
    scores["CBSA_CODE"] = scores["CBSA_CODE"].apply(z5)
    MEFI2017 = scores[scores["year"]==2017].copy()

    # Prepare MEFI columns (your existing code for extracting components)
    MEFI2017_main = (MEFI2017
        .rename(columns={
            "Overall":"MEFI2017_overall",
            "Area1_GovtSpending":"MEFI2017_area1",
            "Area2_Taxes":"MEFI2017_area2",
            "Area3_Labor":"MEFI2017_area3"
        })
        [["CBSA_CODE","MEFI2017_overall","MEFI2017_area1","MEFI2017_area2","MEFI2017_area3"]]
        .drop_duplicates(subset=["CBSA_CODE"])
    )

    # Note: Add your component extraction code here (1A-3C, BEA pop, etc.)
    # This is just the core scores for brevity

    # ========= Merge BRFSS with MEFI =========
    print("\nStep 5: Merging BRFSS with MEFI...")
    BRFSS2017_MEFI = merge_brfss_mefi(B17, MEFI2017_main, "2017")
    BRFSS2022_MEFI = merge_brfss_mefi(B22, MEFI2017_main, "2017")

    # ========= Report results =========
    print("\n" + "=" * 80)
    print("RESULTS SUMMARY")
    print("=" * 80)

    for year, df in [("2017", BRFSS2017_MEFI), ("2022", BRFSS2022_MEFI)]:
        print(f"\n{year}:")
        print(f"  Total rows: {len(df):,}")
        print(f"\n  Join Status:")
        for status, count in df["join_status"].value_counts().items():
            pct = 100 * count / len(df)
            print(f"    {status:15s}: {count:8,} ({pct:5.1f}%)")

        print(f"\n  Exclusion Breakdown:")
        print(f"    Micropolitan:  {(df['is_micro']==1).sum():,}")
        print(f"    PR/Territory:  {(df['is_pr']==1).sum():,}")
        print(f"    Historic MSA:  {(df['is_excluded_historic']==1).sum():,}")
        print(f"    MDivs rolled:  {(df['is_mdiv']==1).sum():,}")

    # ========= Save outputs =========
    print("\n" + "=" * 80)
    print("SAVING REFINED DATA")
    print("=" * 80)

    out_dir = "/content/drive/MyDrive/Saksham_MEFI/"

    BRFSS2017_MEFI.to_csv(out_dir + "BRFSS2017_MEFI_REFINED.csv", index=False)
    BRFSS2022_MEFI.to_csv(out_dir + "BRFSS2022_MEFI_REFINED.csv", index=False)
    BRFSS2017_MEFI.to_parquet(out_dir + "BRFSS2017_MEFI_REFINED.parquet", index=False)
    BRFSS2022_MEFI.to_parquet(out_dir + "BRFSS2022_MEFI_REFINED.parquet", index=False)

    print(f"  ✓ Saved: BRFSS2017_MEFI_REFINED.csv")
    print(f"  ✓ Saved: BRFSS2017_MEFI_REFINED.parquet")
    print(f"  ✓ Saved: BRFSS2022_MEFI_REFINED.csv")
    print(f"  ✓ Saved: BRFSS2022_MEFI_REFINED.parquet")

    # ========= Mystery codes report =========
    print("\n" + "=" * 80)
    print("MYSTERY CODES STILL NEED INVESTIGATION")
    print("=" * 80)

    for code in MYSTERY_CODES_TODO:
        for year, df in [("2017", BRFSS2017_MEFI), ("2022", BRFSS2022_MEFI)]:
            count = (df["CBSA_CODE_ORIG"] == code).sum()
            if count > 0:
                sample = df[df["CBSA_CODE_ORIG"] == code].iloc[0]
                metro_name = sample.get("BRFSS_Metro_Name_Orig", "N/A")
                print(f"\n  Code {code} ({year}): {count:,} rows")
                print(f"    Metro name in BRFSS: {metro_name}")
                print(f"    → TODO: Search OMB for this name manually")

    print("\n" + "=" * 80)
    print("✨ PATCH COMPLETE!")
    print("=" * 80)
    print("\nNext steps:")
    print("  1. Review the join status improvements above")
    print("  2. Investigate the mystery codes manually")
    print("  3. Use BRFSS2017_MEFI_REFINED.csv for your analysis")
````

````python
"""
BRFSS-MEFI DATA REFINEMENT PATCH FILE
Generated: Based on Option 1 & 2 investigation results

This patch updates your existing code with:
1. Comprehensive Metropolitan Division → MSA mappings (13 → 42)
2. Complete Micropolitan areas list (4 → 603)
3. Complete Puerto Rico/Territories list (2 → 14)
4. New columns for MDiv/MSA names
5. Poughkeepsie exclusion
6. TEMPORAL REMAPPINGS for codes that changed over time

EXPECTED IMPACT:
- Match rate: 92% → 97%+
- Unmatched rows: 14,000 → ~3,800
- Fixes 10,000+ rows with DC MDivs and micropolitan codes
"""

import pandas as pd
import numpy as np
import json

# ============================================================================
# SECTION 1: COMPREHENSIVE MAPPING DICTIONARIES
# Replace your existing dictionaries with these
# ============================================================================

# ========== Metropolitan Division → Parent MSA Mapping ==========
# OLD: You had 13 mappings
# NEW: 42 mappings (from OMB List 1)
DIV2CBSA = {
    "11244": "31080",  # Los Angeles MDivs
    "31084": "31080",

    "11694": "47900",  # Washington DC MDivs
    "23224": "47900",
    "43524": "47900",  # Silver Spring-Frederick-Rockville, MD ← CONFIRMED: Associate to Washington MSA
    "47764": "47900",
    "47894": "47900",  # Washington-Arlington-Alexandria, DC-VA-MD-WV ← CONFIRMED: Associate to Washington MSA ← FIXES 8,500 ROWS!

    "12054": "12060",  # Atlanta MDivs
    "31924": "12060",

    "14454": "14460",  # Boston MDivs
    "15764": "14460",
    "40484": "14460",

    "15804": "37980",  # Philadelphia MDivs
    "33874": "37980",
    "37964": "37980",
    "48864": "37980",

    "16974": "16980",  # Chicago MDivs
    "16984": "16980",
    "20994": "16980",
    "23844": "16980",
    "29404": "16980",
    "29414": "16980",

    "19124": "19100",  # Dallas MDivs
    "23104": "19100",

    "19804": "19820",  # Detroit MDivs
    "47664": "19820",

    "20524": "35620",  # New York MDivs
    "29484": "35620",
    "35004": "35620",
    "35084": "35620",
    "35614": "35620",

    "21794": "42660",  # Seattle MDivs
    "42644": "42660",
    "45104": "42660",

    "22744": "33100",  # Miami MDivs
    "33124": "33100",
    "48424": "33100",

    "36084": "41860",  # San Francisco MDivs
    "41884": "41860",
    "42034": "41860",

    "41304": "45300",  # Tampa MDivs
    "45294": "45300",
}

# ========== Legacy Code Renames (Keep your existing ones) ==========
LEGACY_RENAMES = {
    "17200": "30150",  # Claremont-Lebanon → Lebanon-Claremont
    "19430": "19380",  # Dayton-Springfield → Dayton
}

# ========== NEW: Temporal Remappings for Codes That Changed Over Time ==========
# These codes represent the same geographic area but changed codes between years
# Apply AFTER MDiv rollup but BEFORE legacy renames
TEMPORAL_REMAPPINGS = {
    "30100": "30150",  # Lebanon, NH-VT (pre-2013) → Lebanon-Claremont, NH-VT (current)
                       # Note: 30150 is a micropolitan area, so will be excluded
                       # 2,066 rows affected - excluded as micropolitan

    "35154": "35620",  # New Brunswick-Lakewood MDiv (2013-2023) → NY-Newark-Jersey City MSA
                       # Document: OMB Bulletin 23-01 changed 35154→29484, but we roll directly to parent MSA
                       # 2,235 rows affected - JOINS to MEFI via 35620!
}

# ========== Micropolitan Areas (EXCLUDE) ==========
# OLD: You had 4 hardcoded
# NEW: 603 comprehensive (all micros from OMB)
# Note: 30150 is included here, so 30100 → 30150 will be excluded
MICROPOLITAN_AREAS = {
    "10100", "10140", "10220", "10260", "10300", "10460", "10480", "10620",
    "10660", "10700", "10760", "10820", "10860", "10940", "10980", "11060",
    "11140", "11220", "11360", "11380", "11420", "11580", "11620", "11660",
    "11680", "11740", "11780", "11820", "11860", "11900", "11940", "11980",
    "12140", "12180", "12300", "12380", "12460", "12520", "12660", "12680",
    "12740", "12780", "12820", "12860", "12900", "13060", "13100", "13180",
    "13260", "13300", "13340", "13420", "13500", "13540", "13620", "13660",
    "13700", "13720", "13860", "13940", "14100", "14140", "14180", "14220",
    "14300", "14340", "14380", "14420", "14580", "14620", "14660", "14700",
    "14710", "14720", "14780", "14820", "14940", "15020", "15060", "15100",
    "15220", "15340", "15420", "15460", "15580", "15620", "15660", "15700",
    "15740", "15780", "15820", "15860", "15900", "16060", "16100", "16140",
    "16260", "16340", "16380", "16460", "16500", "16660", "17060", "17200",
    "17220", "17260", "17340", "17380", "17500", "17540", "17580", "17620",
    "17640", "17650", "17700", "17740", "18060", "18100", "18180", "18220",
    "18260", "18300", "18340", "18380", "18420", "18460", "18500", "18620",
    "18660", "18740", "18780", "18820", "18860", "18900", "18980", "19000",
    "19060", "19180", "19220", "19260", "19420", "19540", "19580", "19620",
    "19700", "19760", "19810", "19860", "19940", "19980", "20060", "20140",
    "20180", "20300", "20340", "20380", "20420", "20460", "20540", "20580",
    "20660", "20700", "20780", "20820", "20900", "20980", "21020", "21120",
    "21180", "21220", "21260", "21380", "21420", "21460", "21540", "21580",
    "21640", "21700", "21740", "21840", "21860", "21900", "21980", "22060",
    "22100", "22190", "22260", "22280", "22300", "22340", "22580", "22620",
    "22700", "22780", "22800", "22820", "22840", "22860", "23140", "23180",
    "23190", "23240", "23300", "23340", "23380", "23500", "23620", "23660",
    "23680", "23700", "23780", "23820", "23860", "23940", "23980", "24060",
    "24100", "24180", "24330", "24380", "24460", "24600", "24620", "24640",
    "24700", "24740", "24820", "24900", "24940", "24980", "25100", "25200",
    "25300", "25460", "25580", "25700", "25720", "25740", "25760", "25770",
    "25775", "25780", "25820", "25840", "25850", "25880", "25900", "26020",
    "26090", "26220", "26340", "26460", "26500", "26540", "26660", "26700",
    "26740", "26780", "26860", "26940", "26960", "27020", "27160", "27220",
    "27300", "27380", "27420", "27460", "27540", "27580", "27600", "27700",
    "27920", "27940", "28060", "28180", "28260", "28300", "28340", "28380",
    "28500", "28540", "28580", "28620", "28680", "28780", "28820", "28860",
    "28900", "29060", "29260", "29300", "29380", "29470", "29500", "29660",
    "29720", "29780", "29860", "29900", "29980", "30060", "30150", "30180",
    "30220", "30260", "30280", "30380", "30420", "30580", "30660", "30820",
    "30880", "30900", "30940", "31060", "31220", "31260", "31300", "31380",
    "31500", "31580", "31620", "31660", "31680", "31820", "31930", "31940",
    "31980", "32000", "32020", "32060", "32100", "32140", "32180", "32220",
    "32260", "32280", "32300", "32340", "32380", "32390", "32460", "32500",
    "32540", "32620", "32660", "32700", "32740", "32860", "32940", "32980",
    "33020", "33060", "33180", "33300", "33380", "33420", "33500", "33580",
    "33620", "33900", "33910", "33940", "33980", "34020", "34140", "34180",
    "34220", "34260", "34300", "34340", "34380", "34420", "34460", "34500",
    "34540", "34660", "34680", "34700", "34780", "34860", "34880", "35020",
    "35060", "35100", "35140", "35220", "35260", "35340", "35420", "35440",
    "35460", "35500", "35580", "35700", "35740", "35820", "35860", "35900",
    "35940", "36020", "36180", "36300", "36340", "36380", "36460", "36580",
    "36620", "36660", "36700", "36820", "36830", "36837", "36840", "36860",
    "36900", "36940", "37020", "37060", "37080", "37120", "37140", "37220",
    "37260", "37300", "37420", "37500", "37540", "37580", "37660", "37740",
    "37780", "37800", "37940", "37950", "38100", "38180", "38210", "38220",
    "38240", "38260", "38380", "38420", "38460", "38500", "38580", "38620",
    "38700", "38740", "38780", "38820", "38840", "38920", "39020", "39040",
    "39060", "39220", "39260", "39420", "39480", "39500", "39700", "39780",
    "39860", "39940", "39960", "39980", "40080", "40090", "40100", "40180",
    "40260", "40300", "40460", "40540", "40620", "40700", "40740", "40760",
    "40770", "40780", "40820", "40860", "40940", "41220", "41260", "41400",
    "41460", "41760", "41780", "41820", "42180", "42300", "42380", "42420",
    "42460", "42580", "42620", "42740", "42780", "42820", "42860", "42900",
    "42940", "42980", "43020", "43060", "43140", "43180", "43220", "43260",
    "43320", "43380", "43460", "43500", "43660", "43700", "43740", "43760",
    "43890", "43940", "43980", "44020", "44260", "44340", "44460", "44500",
    "44540", "44580", "44620", "44660", "44740", "44780", "44860", "44900",
    "44920", "44980", "45000", "45020", "45140", "45180", "45340", "45380",
    "45520", "45580", "45620", "45660", "45700", "45740", "45860", "45880",
    "45900", "45980", "46020", "46100", "46180", "46300", "46380", "46460",
    "46500", "46580", "46620", "46740", "46780", "46820", "46860", "46900",
    "46980", "47080", "47180", "47240", "47340", "47420", "47540", "47620",
    "47660", "47700", "47780", "47820", "47920", "47980", "48020", "48100",
    "48180", "48200", "48220", "48460", "48580", "48780", "48820", "48940",
    "48980", "49010", "49080", "49100", "49220", "49260", "49300", "49380",
    "49460", "49780", "49820"
}

# ========== Puerto Rico & Territories (EXCLUDE) ==========
# OLD: You had 2 hardcoded
# NEW: 14 comprehensive (all PR metros from OMB)
PUERTO_RICO_TERRITORIES = {
    "10260",  # Adjuntas, PR
    "10380",  # Aguadilla, PR
    "11640",  # Arecibo, PR
    "17620",  # Coamo, PR
    "17640",  # Coco, PR
    "25020",  # Guayama, PR
    "27580",  # Jayuya, PR
    "29720",  # Lares, PR
    "32420",  # Mayagüez, PR
    "38660",  # Ponce, PR
    "41900",  # San Germán, PR
    "41980",  # San Juan-Bayamón-Caguas, PR
    "42180",  # Santa Isabel, PR
    "46580",  # Utuado, PR
}

# ========== Historic MSAs to EXCLUDE ==========
# NEW: Based on investigation, Poughkeepsie should be excluded
# 39100: 517 rows, appears to be Poughkeepsie - CONFIRMED: exclude
EXCLUDED_HISTORIC_MSAS = {
    "39100",  # Poughkeepsie-Newburgh-Middletown, NY (appears in 2022, not in MEFI properly)
              # 517 rows affected - excluded per protocol
}

# ========== Mystery Codes - NOW RESOLVED! ==========
# These codes have been investigated and resolved via TEMPORAL_REMAPPINGS
# OLD MYSTERY_CODES_TODO (now handled):
#   "35154" → now mapped to "35620" via TEMPORAL_REMAPPINGS (joins to MEFI!)
#   "30100" → now mapped to "30150" via TEMPORAL_REMAPPINGS (excluded as micropolitan)
MYSTERY_CODES_TODO = set()  # All mystery codes have been resolved!


# ============================================================================
# SECTION 2: UPDATED HELPER FUNCTIONS
# Keep your existing z5() and norm_name(), add these new ones
# ============================================================================

def z5(x):
    """Pad CBSA codes to 5 digits with leading zeros"""
    if pd.isna(x): return pd.NA
    s = str(x).strip()
    if s == "" or s.lower() == "nan": return pd.NA
    if s.replace(".","",1).isdigit():
        s = str(int(float(s)))
    return s.zfill(5)

def norm_name(s):
    """Normalize metro area names for matching"""
    if pd.isna(s): return ""
    s = str(s).lower()
    for ch in ["–","—","(",")",".",","]:
        s = s.replace(ch, " ")
    s = s.replace("-", " ")
    return " ".join(s.split())

def add_scope_flags(df, code_col="CBSA_CODE_CANON"):
    """
    Add comprehensive exclusion flags
    UPDATED: Now uses comprehensive sets from OMB extraction
    """
    df["is_micro"] = df[code_col].isin(MICROPOLITAN_AREAS).astype(int)
    df["is_pr"] = df[code_col].isin(PUERTO_RICO_TERRITORIES).astype(int)
    df["is_excluded_historic"] = df[code_col].isin(EXCLUDED_HISTORIC_MSAS).astype(int)

    def _note(code):
        if code in MICROPOLITAN_AREAS:
            return "Micropolitan area — MEFI not available"
        if code in PUERTO_RICO_TERRITORIES:
            return "Puerto Rico/Territory — MEFI typically excluded"
        if code in EXCLUDED_HISTORIC_MSAS:
            return "Historic MSA in MEFI — excluded per protocol"
        if code in MYSTERY_CODES_TODO:
            return "Under investigation — code not in OMB"
        return ""

    df["scope_note"] = df[code_col].map(_note)

    # Master exclusion flag
    df["exclude_from_analysis"] = (
        (df["is_micro"] == 1) |
        (df["is_pr"] == 1) |
        (df["is_excluded_historic"] == 1)
    ).astype(int)

    return df


# ============================================================================
# SECTION 3: UPDATED prep_brfss FUNCTION
# Replace your existing prep_brfss function with this
# ============================================================================

def prep_brfss(df):
    """
    Prepare BRFSS data with comprehensive CBSA code standardization

    UPDATED: Now includes:
    - Comprehensive MDiv rollup (42 mappings)
    - Temporal remappings for codes that changed over time
    - MDiv name retention
    - MSA name lookup
    - Comprehensive exclusion flags
    """
    df = df.copy()
    assert "CBSA_CODE" in df.columns, "CBSA_CODE missing in BRFSS file."

    # Step 1: Standardize original code
    df["CBSA_CODE_ORIG"] = df["CBSA_CODE"].apply(z5)

    # Step 2: Roll up Metropolitan Divisions to parent MSAs
    df["CBSA_CODE_ROLLED"] = df["CBSA_CODE_ORIG"].map(DIV2CBSA).fillna(df["CBSA_CODE_ORIG"])

    # Step 3: NEW - Apply temporal remappings for codes that changed over time
    # This handles cases like:
    #   - 30100 (Lebanon pre-2013) → 30150 (Lebanon-Claremont current)
    #   - 35154 (New Brunswick MDiv 2013-2023) → 35620 (NY-Newark MSA)
    df["CBSA_CODE_TEMPORAL"] = df["CBSA_CODE_ROLLED"].replace(TEMPORAL_REMAPPINGS)

    # Step 4: Apply legacy renames (e.g., Dayton, other historical changes)
    df["CBSA_CODE_CANON"] = df["CBSA_CODE_TEMPORAL"].replace(LEGACY_RENAMES)

    # Step 5: Flag if original was a Metropolitan Division
    df["is_mdiv"] = df["CBSA_CODE_ORIG"].isin(DIV2CBSA.keys()).astype(int)

    # Step 6: Flag if code was temporally remapped
    df["is_temporal_remap"] = df["CBSA_CODE_ORIG"].isin(TEMPORAL_REMAPPINGS.keys()).astype(int)

    # Step 7: Add comprehensive exclusion flags
    df = add_scope_flags(df, "CBSA_CODE_CANON")

    # Step 8: Store original metro name from BRFSS
    if "MMSANAME" in df.columns:
        df["BRFSS_Metro_Name_Orig"] = df["MMSANAME"]

    # Step 9: Drop original CBSA_CODE to avoid confusion
    df = df.drop(columns=["CBSA_CODE"], errors="ignore")

    return df


# ============================================================================
# SECTION 4: ADD MSA/MDIV NAMES FROM OMB
# NEW: This adds the name columns you requested
# ============================================================================

def add_msa_names_from_omb(brfss_df, omb_2023_path):
    """
    Add MSA names from OMB List 1 (2023) to BRFSS data
    Creates columns: MDiv_Name, MSA_Name_Current

    Args:
        brfss_df: BRFSS DataFrame (after prep_brfss)
        omb_2023_path: Path to list1_2023.xlsx file

    Returns:
        DataFrame with added name columns
    """
    # Load OMB 2023
    omb = pd.read_excel(omb_2023_path, sheet_name="List 1", dtype=str, header=2)
    omb.columns = [str(c).strip() for c in omb.columns]

    # Standardize codes
    if "CBSA Code" in omb.columns:
        omb["CBSA Code"] = omb["CBSA Code"].apply(z5)
    if "Metropolitan Division Code" in omb.columns:
        omb["Metropolitan Division Code"] = omb["Metropolitan Division Code"].apply(z5)

    # Create MDiv name lookup
    mdiv_names = {}
    if "Metropolitan Division Code" in omb.columns and "Metropolitan Division Title" in omb.columns:
        mdiv_data = omb[["Metropolitan Division Code", "Metropolitan Division Title"]].dropna()
        mdiv_data = mdiv_data.drop_duplicates(subset=["Metropolitan Division Code"])
        mdiv_names = dict(mdiv_data.values)

    # Create CBSA name lookup (for parent MSAs)
    cbsa_names = {}
    if "CBSA Code" in omb.columns and "CBSA Title" in omb.columns:
        cbsa_data = omb[["CBSA Code", "CBSA Title"]].dropna()
        cbsa_data = cbsa_data.drop_duplicates(subset=["CBSA Code"])
        cbsa_names = dict(cbsa_data.values)

    # Add MDiv name if original code was a MDiv
    brfss_df["MDiv_Name"] = brfss_df.apply(
        lambda row: mdiv_names.get(row["CBSA_CODE_ORIG"]) if row["is_mdiv"] == 1 else pd.NA,
        axis=1
    )

    # Add current MSA name (from canonical code)
    brfss_df["MSA_Name_Current"] = brfss_df["CBSA_CODE_CANON"].map(cbsa_names)

    return brfss_df


# ============================================================================
# SECTION 5: UPDATED MERGE FUNCTION
# Replace your existing merge_brfss_mefi function with this
# ============================================================================

def merge_brfss_mefi(brfss_df, mefi_df, year_label="2017"):
    """
    Merge BRFSS data with MEFI scores and add comprehensive join status flags

    UPDATED: Now handles exclusions and mystery codes properly
    """
    out = brfss_df.merge(mefi_df, left_on="CBSA_CODE_CANON", right_on="CBSA_CODE", how="left")

    mefi_core = [f"MEFI{year_label}_overall", f"MEFI{year_label}_area1",
                 f"MEFI{year_label}_area2", f"MEFI{year_label}_area3"]
    mefi_core_existing = [c for c in mefi_core if c in out.columns]

    out["mefi_joined"] = out[mefi_core_existing].notna().any(axis=1).astype(int)

    # Updated join_status to handle exclusions
    out["join_status"] = np.where(
        out["mefi_joined"].eq(1),
        "joined",
        np.where(
            out["exclude_from_analysis"].eq(1),
            "excluded",
            "unmatched"
        )
    )

    # Updated mefi_join_note with more detail
    out["mefi_join_note"] = np.where(
        out["mefi_joined"].eq(1),
        "MEFI attached",
        np.where(
            out["scope_note"].ne(""),
            out["scope_note"],
            "No MEFI match — check mapping"
        )
    )

    return out


# ============================================================================
# SECTION 6: MAIN EXECUTION EXAMPLE
# Replace your main execution block with this pattern
# ============================================================================

if __name__ == "__main__":

    print("=" * 80)
    print("APPLYING COMPREHENSIVE BRFSS-MEFI REFINEMENT PATCH")
    print("WITH TEMPORAL REMAPPINGS")
    print("=" * 80)

    # ========= File paths (UPDATE THESE) =========
    SMART2017 = "/content/drive/MyDrive/Saksham_MEFI/SMART_2017_individual.csv"
    SMART2022 = "/content/drive/MyDrive/Saksham_MEFI/SMART_2022_individual.csv"
    MEFI_XLSX = "/content/drive/MyDrive/Saksham_MEFI/MEFI+2022+dataset+for+researchers,+1-5-23 (4).xlsx"
    OMB_2023 = "/content/drive/MyDrive/Saksham_MEFI/list1_2023.xlsx"

    # ========= Load BRFSS data =========
    print("\nStep 1: Loading BRFSS data...")
    B17 = pd.read_csv(SMART2017, dtype=str, low_memory=False)
    B22 = pd.read_csv(SMART2022, dtype=str, low_memory=False)
    print(f"  2017: {len(B17):,} rows")
    print(f"  2022: {len(B22):,} rows")

    # ========= Apply comprehensive prep =========
    print("\nStep 2: Applying comprehensive CBSA standardization...")
    B17 = prep_brfss(B17)
    B22 = prep_brfss(B22)
    print("  ✓ Applied MDiv rollup (42 mappings)")
    print("  ✓ Applied temporal remappings (2 codes: 30100→30150, 35154→35620)")
    print("  ✓ Applied legacy renames")
    print("  ✓ Applied exclusion flags (603 micros, 14 PR/territories, 1 historic)")

    # ========= Add MSA/MDiv names =========
    print("\nStep 3: Adding MSA/MDiv names from OMB...")
    B17 = add_msa_names_from_omb(B17, OMB_2023)
    B22 = add_msa_names_from_omb(B22, OMB_2023)
    print("  ✓ Added MDiv_Name column")
    print("  ✓ Added MSA_Name_Current column")

    # ========= Load and prep MEFI data =========
    print("\nStep 4: Loading MEFI data...")
    scores = pd.read_excel(MEFI_XLSX, sheet_name="scores (panel)", header=0)
    scores = scores.rename(columns={
        "CBSA Code":"CBSA_CODE",
        "Area 1: Govern-ment Spending":"Area1_GovtSpending",
        "Area 2: Taxes":"Area2_Taxes",
        "Area 3: Labor Market Regulation":"Area3_Labor",
    })
    scores["CBSA_CODE"] = scores["CBSA_CODE"].apply(z5)
    MEFI2017 = scores[scores["year"]==2017].copy()

    # Prepare MEFI columns (your existing code for extracting components)
    MEFI2017_main = (MEFI2017
        .rename(columns={
            "Overall":"MEFI2017_overall",
            "Area1_GovtSpending":"MEFI2017_area1",
            "Area2_Taxes":"MEFI2017_area2",
            "Area3_Labor":"MEFI2017_area3"
        })
        [["CBSA_CODE","MEFI2017_overall","MEFI2017_area1","MEFI2017_area2","MEFI2017_area3"]]
        .drop_duplicates(subset=["CBSA_CODE"])
    )

    # Note: Add your component extraction code here (1A-3C, BEA pop, etc.)
    # This is just the core scores for brevity

    # ========= Merge BRFSS with MEFI =========
    print("\nStep 5: Merging BRFSS with MEFI...")
    BRFSS2017_MEFI = merge_brfss_mefi(B17, MEFI2017_main, "2017")
    BRFSS2022_MEFI = merge_brfss_mefi(B22, MEFI2017_main, "2017")

    # ========= Report results =========
    print("\n" + "=" * 80)
    print("RESULTS SUMMARY")
    print("=" * 80)

    for year, df in [("2017", BRFSS2017_MEFI), ("2022", BRFSS2022_MEFI)]:
        print(f"\n{year}:")
        print(f"  Total rows: {len(df):,}")
        print(f"\n  Join Status:")
        for status, count in df["join_status"].value_counts().items():
            pct = 100 * count / len(df)
            print(f"    {status:15s}: {count:8,} ({pct:5.1f}%)")

        print(f"\n  Exclusion Breakdown:")
        print(f"    Micropolitan:      {(df['is_micro']==1).sum():,}")
        print(f"    PR/Territory:      {(df['is_pr']==1).sum():,}")
        print(f"    Historic MSA:      {(df['is_excluded_historic']==1).sum():,}")
        print(f"    MDivs rolled:      {(df['is_mdiv']==1).sum():,}")
        print(f"    Temporal remaps:   {(df['is_temporal_remap']==1).sum():,}")

    # ========= Temporal remapping report =========
    print("\n" + "=" * 80)
    print("TEMPORAL REMAPPING SUMMARY")
    print("=" * 80)
    print("\nThe following codes were remapped due to temporal changes:")
    for old_code, new_code in TEMPORAL_REMAPPINGS.items():
        for year, df in [("2017", BRFSS2017_MEFI), ("2022", BRFSS2022_MEFI)]:
            count = (df["CBSA_CODE_ORIG"] == old_code).sum()
            if count > 0:
                is_excluded = new_code in MICROPOLITAN_AREAS
                status = "excluded (micropolitan)" if is_excluded else "joined to MEFI"
                print(f"  {old_code} → {new_code} ({year}): {count:,} rows — {status}")

    # ========= Save outputs =========
    print("\n" + "=" * 80)
    print("SAVING REFINED DATA")
    print("=" * 80)

    out_dir = "/content/drive/MyDrive/Saksham_MEFI/"

    BRFSS2017_MEFI.to_csv(out_dir + "BRFSS2017_MEFI_REFINED1.csv", index=False)
    BRFSS2022_MEFI.to_csv(out_dir + "BRFSS2022_MEFI_REFINED1.csv", index=False)
    BRFSS2017_MEFI.to_parquet(out_dir + "BRFSS2017_MEFI_REFINED1.parquet", index=False)
    BRFSS2022_MEFI.to_parquet(out_dir + "BRFSS2022_MEFI_REFINED1.parquet", index=False)

    print(f"  ✓ Saved: BRFSS2017_MEFI_REFINED1.csv")
    print(f"  ✓ Saved: BRFSS2017_MEFI_REFINED1.parquet")
    print(f"  ✓ Saved: BRFSS2022_MEFI_REFINED1.csv")
    print(f"  ✓ Saved: BRFSS2022_MEFI_REFINED1.parquet")

    # ========= Mystery codes report =========
    print("\n" + "=" * 80)
    print("MYSTERY CODES STATUS")
    print("=" * 80)

    if len(MYSTERY_CODES_TODO) == 0:
        print("\n  ✨ All mystery codes have been resolved!")
        print("  • 30100 → remapped to 30150 (micropolitan, excluded)")
        print("  • 35154 → remapped to 35620 (joins to MEFI!)")
        print("  • 39100 → Poughkeepsie (excluded)")
        print("  • 43524 → Washington DC MDiv (joins to MEFI)")
        print("  • 47894 → Washington DC MDiv (joins to MEFI)")
    else:
        print("\n  Remaining mystery codes:")
        for code in MYSTERY_CODES_TODO:
            for year, df in [("2017", BRFSS2017_MEFI), ("2022", BRFSS2022_MEFI)]:
                count = (df["CBSA_CODE_ORIG"] == code).sum()
                if count > 0:
                    sample = df[df["CBSA_CODE_ORIG"] == code].iloc[0]
                    metro_name = sample.get("BRFSS_Metro_Name_Orig", "N/A")
                    print(f"\n  Code {code} ({year}): {count:,} rows")
                    print(f"    Metro name in BRFSS: {metro_name}")
                    print(f"    → TODO: Search OMB for this name manually")

    print("\n" + "=" * 80)
    print("✨ PATCH COMPLETE!")
    print("=" * 80)
    print("\nNext steps:")
    print("  1. Review the join status improvements above")
    print("  2. All mystery codes have been resolved!")
    print("  3. Use BRFSS2017_MEFI_REFINED.csv for your analysis")
    print("  4. Expected improvements:")
    print("     • 30100 (2,066 rows) → excluded as micropolitan")
    print("     • 35154 (2,235 rows) → JOINED to MEFI via 35620!")
    print("     • 39100 (517 rows) → excluded (Poughkeepsie)")
````

````python
"""
BRFSS-MEFI DATA REFINEMENT PATCH FILE - ENHANCED EDITION
Generated: Based on Option 1 & 2 investigation results + Component scores + Clean outputs

This patch updates your existing code with:
1. Comprehensive Metropolitan Division → MSA mappings (13 → 42)
2. Complete Micropolitan areas list (4 → 603)
3. Complete Puerto Rico/Territories list (2 → 14)
4. New columns for MDiv/MSA names
5. Poughkeepsie exclusion
6. TEMPORAL REMAPPINGS for codes that changed over time (30100, 35154)
7. Component scores extraction (1A-3C) from individual sheets
8. BEA population data (2017)
9. Clean and semi-clean output files

EXPECTED IMPACT:
- Match rate: 92% → 97%+
- Unmatched rows: 14,000 → ~3,800
- Fixes 10,000+ rows with DC MDivs and micropolitan codes
- Includes all area scores (1, 2, 3) and component scores (1A-3C)
- Includes BEA population data

OUTPUT FILES PER YEAR:
1. BRFSS{year}_MEFI_CLEAN.csv - Only successfully joined rows
2. BRFSS{year}_MEFI_SEMICLEAN.csv - Joined + unmatched (excludes micro/PR/historic)
3. BRFSS{year}_MEFI_FULL.csv - All rows with flags for transparency
"""

import pandas as pd
import numpy as np
import json

# ============================================================================
# SECTION 1: COMPREHENSIVE MAPPING DICTIONARIES
# Replace your existing dictionaries with these
# ============================================================================

# ========== Metropolitan Division → Parent MSA Mapping ==========
# OLD: You had 13 mappings
# NEW: 42 mappings (from OMB List 1)
DIV2CBSA = {
    "11244": "31080",  # Los Angeles MDivs
    "31084": "31080",

    "11694": "47900",  # Washington DC MDivs
    "23224": "47900",
    "43524": "47900",  # Silver Spring-Frederick-Rockville, MD ← CONFIRMED: Associate to Washington MSA
    "47764": "47900",
    "47894": "47900",  # Washington-Arlington-Alexandria, DC-VA-MD-WV ← CONFIRMED: Associate to Washington MSA ← FIXES 8,500 ROWS!

    "12054": "12060",  # Atlanta MDivs
    "31924": "12060",

    "14454": "14460",  # Boston MDivs
    "15764": "14460",
    "40484": "14460",

    "15804": "37980",  # Philadelphia MDivs
    "33874": "37980",
    "37964": "37980",
    "48864": "37980",

    "16974": "16980",  # Chicago MDivs
    "16984": "16980",
    "20994": "16980",
    "23844": "16980",
    "29404": "16980",
    "29414": "16980",

    "19124": "19100",  # Dallas MDivs
    "23104": "19100",

    "19804": "19820",  # Detroit MDivs
    "47664": "19820",

    "20524": "35620",  # New York MDivs
    "29484": "35620",
    "35004": "35620",
    "35084": "35620",
    "35614": "35620",

    "21794": "42660",  # Seattle MDivs
    "42644": "42660",
    "45104": "42660",

    "22744": "33100",  # Miami MDivs
    "33124": "33100",
    "48424": "33100",

    "36084": "41860",  # San Francisco MDivs
    "41884": "41860",
    "42034": "41860",

    "41304": "45300",  # Tampa MDivs
    "45294": "45300",
}

# ========== Legacy Code Renames (Keep your existing ones) ==========
LEGACY_RENAMES = {
    "17200": "30150",  # Claremont-Lebanon → Lebanon-Claremont
    "19430": "19380",  # Dayton-Springfield → Dayton
}

# ========== NEW: Temporal Remappings for Codes That Changed Over Time ==========
# These codes represent the same geographic area but changed codes between years
# Apply AFTER MDiv rollup but BEFORE legacy renames
TEMPORAL_REMAPPINGS = {
    "30100": "30150",  # Lebanon, NH-VT (pre-2013) → Lebanon-Claremont, NH-VT (current)
                       # Note: 30150 is a micropolitan area, so will be excluded
                       # 2,066 rows affected - excluded as micropolitan

    "35154": "35620",  # New Brunswick-Lakewood MDiv (2013-2023) → NY-Newark-Jersey City MSA
                       # Document: OMB Bulletin 23-01 changed 35154→29484, but we roll directly to parent MSA
                       # 2,235 rows affected - JOINS to MEFI via 35620!
}

# ========== Micropolitan Areas (EXCLUDE) ==========
# OLD: You had 4 hardcoded
# NEW: 603 comprehensive (all micros from OMB)
# Note: 30150 is included here, so 30100 → 30150 will be excluded
MICROPOLITAN_AREAS = {
    "10100", "10140", "10220", "10260", "10300", "10460", "10480", "10620",
    "10660", "10700", "10760", "10820", "10860", "10940", "10980", "11060",
    "11140", "11220", "11360", "11380", "11420", "11580", "11620", "11660",
    "11680", "11740", "11780", "11820", "11860", "11900", "11940", "11980",
    "12140", "12180", "12300", "12380", "12460", "12520", "12660", "12680",
    "12740", "12780", "12820", "12860", "12900", "13060", "13100", "13180",
    "13260", "13300", "13340", "13420", "13500", "13540", "13620", "13660",
    "13700", "13720", "13860", "13940", "14100", "14140", "14180", "14220",
    "14300", "14340", "14380", "14420", "14580", "14620", "14660", "14700",
    "14710", "14720", "14780", "14820", "14940", "15020", "15060", "15100",
    "15220", "15340", "15420", "15460", "15580", "15620", "15660", "15700",
    "15740", "15780", "15820", "15860", "15900", "16060", "16100", "16140",
    "16260", "16340", "16380", "16460", "16500", "16660", "17060", "17200",
    "17220", "17260", "17340", "17380", "17500", "17540", "17580", "17620",
    "17640", "17650", "17700", "17740", "18060", "18100", "18180", "18220",
    "18260", "18300", "18340", "18380", "18420", "18460", "18500", "18620",
    "18660", "18740", "18780", "18820", "18860", "18900", "18980", "19000",
    "19060", "19180", "19220", "19260", "19420", "19540", "19580", "19620",
    "19700", "19760", "19810", "19860", "19940", "19980", "20060", "20140",
    "20180", "20300", "20340", "20380", "20420", "20460", "20540", "20580",
    "20660", "20700", "20780", "20820", "20900", "20980", "21020", "21120",
    "21180", "21220", "21260", "21380", "21420", "21460", "21540", "21580",
    "21640", "21700", "21740", "21840", "21860", "21900", "21980", "22060",
    "22100", "22190", "22260", "22280", "22300", "22340", "22580", "22620",
    "22700", "22780", "22800", "22820", "22840", "22860", "23140", "23180",
    "23190", "23240", "23300", "23340", "23380", "23500", "23620", "23660",
    "23680", "23700", "23780", "23820", "23860", "23940", "23980", "24060",
    "24100", "24180", "24330", "24380", "24460", "24600", "24620", "24640",
    "24700", "24740", "24820", "24900", "24940", "24980", "25100", "25200",
    "25300", "25460", "25580", "25700", "25720", "25740", "25760", "25770",
    "25775", "25780", "25820", "25840", "25850", "25880", "25900", "26020",
    "26090", "26220", "26340", "26460", "26500", "26540", "26660", "26700",
    "26740", "26780", "26860", "26940", "26960", "27020", "27160", "27220",
    "27300", "27380", "27420", "27460", "27540", "27580", "27600", "27700",
    "27920", "27940", "28060", "28180", "28260", "28300", "28340", "28380",
    "28500", "28540", "28580", "28620", "28680", "28780", "28820", "28860",
    "28900", "29060", "29260", "29300", "29380", "29470", "29500", "29660",
    "29720", "29780", "29860", "29900", "29980", "30060", "30150", "30180",
    "30220", "30260", "30280", "30380", "30420", "30580", "30660", "30820",
    "30880", "30900", "30940", "31060", "31220", "31260", "31300", "31380",
    "31500", "31580", "31620", "31660", "31680", "31820", "31930", "31940",
    "31980", "32000", "32020", "32060", "32100", "32140", "32180", "32220",
    "32260", "32280", "32300", "32340", "32380", "32390", "32460", "32500",
    "32540", "32620", "32660", "32700", "32740", "32860", "32940", "32980",
    "33020", "33060", "33180", "33300", "33380", "33420", "33500", "33580",
    "33620", "33900", "33910", "33940", "33980", "34020", "34140", "34180",
    "34220", "34260", "34300", "34340", "34380", "34420", "34460", "34500",
    "34540", "34660", "34680", "34700", "34780", "34860", "34880", "35020",
    "35060", "35100", "35140", "35220", "35260", "35340", "35420", "35440",
    "35460", "35500", "35580", "35700", "35740", "35820", "35860", "35900",
    "35940", "36020", "36180", "36300", "36340", "36380", "36460", "36580",
    "36620", "36660", "36700", "36820", "36830", "36837", "36840", "36860",
    "36900", "36940", "37020", "37060", "37080", "37120", "37140", "37220",
    "37260", "37300", "37420", "37500", "37540", "37580", "37660", "37740",
    "37780", "37800", "37940", "37950", "38100", "38180", "38210", "38220",
    "38240", "38260", "38380", "38420", "38460", "38500", "38580", "38620",
    "38700", "38740", "38780", "38820", "38840", "38920", "39020", "39040",
    "39060", "39220", "39260", "39420", "39480", "39500", "39700", "39780",
    "39860", "39940", "39960", "39980", "40080", "40090", "40100", "40180",
    "40260", "40300", "40460", "40540", "40620", "40700", "40740", "40760",
    "40770", "40780", "40820", "40860", "40940", "41220", "41260", "41400",
    "41460", "41760", "41780", "41820", "42180", "42300", "42380", "42420",
    "42460", "42580", "42620", "42740", "42780", "42820", "42860", "42900",
    "42940", "42980", "43020", "43060", "43140", "43180", "43220", "43260",
    "43320", "43380", "43460", "43500", "43660", "43700", "43740", "43760",
    "43890", "43940", "43980", "44020", "44260", "44340", "44460", "44500",
    "44540", "44580", "44620", "44660", "44740", "44780", "44860", "44900",
    "44920", "44980", "45000", "45020", "45140", "45180", "45340", "45380",
    "45520", "45580", "45620", "45660", "45700", "45740", "45860", "45880",
    "45900", "45980", "46020", "46100", "46180", "46300", "46380", "46460",
    "46500", "46580", "46620", "46740", "46780", "46820", "46860", "46900",
    "46980", "47080", "47180", "47240", "47340", "47420", "47540", "47620",
    "47660", "47700", "47780", "47820", "47920", "47980", "48020", "48100",
    "48180", "48200", "48220", "48460", "48580", "48780", "48820", "48940",
    "48980", "49010", "49080", "49100", "49220", "49260", "49300", "49380",
    "49460", "49780", "49820"
}

# ========== Puerto Rico & Territories (EXCLUDE) ==========
# OLD: You had 2 hardcoded
# NEW: 14 comprehensive (all PR metros from OMB)
PUERTO_RICO_TERRITORIES = {
    "10260",  # Adjuntas, PR
    "10380",  # Aguadilla, PR
    "11640",  # Arecibo, PR
    "17620",  # Coamo, PR
    "17640",  # Coco, PR
    "25020",  # Guayama, PR
    "27580",  # Jayuya, PR
    "29720",  # Lares, PR
    "32420",  # Mayagüez, PR
    "38660",  # Ponce, PR
    "41900",  # San Germán, PR
    "41980",  # San Juan-Bayamón-Caguas, PR
    "42180",  # Santa Isabel, PR
    "46580",  # Utuado, PR
}

# ========== Historic MSAs to EXCLUDE ==========
# NEW: Based on investigation, Poughkeepsie should be excluded
# 39100: 517 rows, appears to be Poughkeepsie - CONFIRMED: exclude
EXCLUDED_HISTORIC_MSAS = {
    "39100",  # Poughkeepsie-Newburgh-Middletown, NY (appears in 2022, not in MEFI properly)
              # 517 rows affected - excluded per protocol
}

# ========== Mystery Codes - NOW RESOLVED! ==========
# These codes have been investigated and resolved via TEMPORAL_REMAPPINGS
# OLD MYSTERY_CODES_TODO (now handled):
#   "35154" → now mapped to "35620" via TEMPORAL_REMAPPINGS (joins to MEFI!)
#   "30100" → now mapped to "30150" via TEMPORAL_REMAPPINGS (excluded as micropolitan)
MYSTERY_CODES_TODO = set()  # All mystery codes have been resolved!


# ============================================================================
# SECTION 2: UPDATED HELPER FUNCTIONS
# Keep your existing z5() and norm_name(), add these new ones
# ============================================================================

def z5(x):
    """Pad CBSA codes to 5 digits with leading zeros"""
    if pd.isna(x): return pd.NA
    s = str(x).strip()
    if s == "" or s.lower() == "nan": return pd.NA
    if s.replace(".","",1).isdigit():
        s = str(int(float(s)))
    return s.zfill(5)

def norm_name(s):
    """Normalize metro area names for matching"""
    if pd.isna(s): return ""
    s = str(s).lower()
    for ch in ["–","—","(",")",".",","]:
        s = s.replace(ch, " ")
    s = s.replace("-", " ")
    return " ".join(s.split())

def pull_component_by_name(xlsx_path, sheet_name, component_label, year=2017):
    """
    Extract a component score (e.g. 1A-3C) from MEFI workbook
    Handles two-row headers and joins by normalized metro area name
    """
    try:
        # Read with two-row header
        df = pd.read_excel(xlsx_path, sheet_name=sheet_name, header=[0,1])

        # Find metro area name column
        name_col = None
        for col in df.columns:
            if isinstance(col, tuple):
                col_combined = ' '.join([str(c).lower() for c in col])
                for part in col:
                    if "metro area" in str(part).lower() or "metro area" in col_combined:
                        name_col = col
                        break
            else:
                if "metro area" in str(col).lower():
                    name_col = col
                    break
            if name_col:
                break

        # Find the score column for the specified year
        score_col = None
        for col in df.columns:
            if isinstance(col, tuple):
                col_combined = ' '.join([str(c).lower() for c in col])
                if str(year) in col_combined:
                    score_col = col
                    break

        if name_col is None or score_col is None:
            print(f"  Warning: Could not find columns in sheet {sheet_name}")
            return pd.DataFrame(columns=["Metro_Area_Name_N", f"MEFI2017_{component_label}"])

        # Extract and clean
        result = df[[name_col, score_col]].copy()
        result.columns = ["Metro_Area_Name", f"MEFI2017_{component_label}"]
        result = result[pd.to_numeric(result[f"MEFI2017_{component_label}"], errors="coerce").notna()].copy()
        result[f"MEFI2017_{component_label}"] = result[f"MEFI2017_{component_label}"].astype(float)
        result["Metro_Area_Name_N"] = result["Metro_Area_Name"].apply(norm_name)

        return result[["Metro_Area_Name_N", f"MEFI2017_{component_label}"]]

    except Exception as e:
        print(f"  Warning: Could not extract {component_label} from {sheet_name}: {e}")
        return pd.DataFrame(columns=["Metro_Area_Name_N", f"MEFI2017_{component_label}"])


def add_scope_flags(df, code_col="CBSA_CODE_CANON"):
    """
    Add comprehensive exclusion flags
    UPDATED: Now uses comprehensive sets from OMB extraction
    """
    df["is_micro"] = df[code_col].isin(MICROPOLITAN_AREAS).astype(int)
    df["is_pr"] = df[code_col].isin(PUERTO_RICO_TERRITORIES).astype(int)
    df["is_excluded_historic"] = df[code_col].isin(EXCLUDED_HISTORIC_MSAS).astype(int)

    def _note(code):
        if code in MICROPOLITAN_AREAS:
            return "Micropolitan area — MEFI not available"
        if code in PUERTO_RICO_TERRITORIES:
            return "Puerto Rico/Territory — MEFI typically excluded"
        if code in EXCLUDED_HISTORIC_MSAS:
            return "Historic MSA in MEFI — excluded per protocol"
        if code in MYSTERY_CODES_TODO:
            return "Under investigation — code not in OMB"
        return ""

    df["scope_note"] = df[code_col].map(_note)

    # Master exclusion flag
    df["exclude_from_analysis"] = (
        (df["is_micro"] == 1) |
        (df["is_pr"] == 1) |
        (df["is_excluded_historic"] == 1)
    ).astype(int)

    return df


# ============================================================================
# SECTION 3: UPDATED prep_brfss FUNCTION
# Replace your existing prep_brfss function with this
# ============================================================================

def prep_brfss(df):
    """
    Prepare BRFSS data with comprehensive CBSA code standardization

    UPDATED: Now includes:
    - Comprehensive MDiv rollup (42 mappings)
    - Temporal remappings for codes that changed over time
    - MDiv name retention
    - MSA name lookup
    - Comprehensive exclusion flags
    """
    df = df.copy()
    assert "CBSA_CODE" in df.columns, "CBSA_CODE missing in BRFSS file."

    # Step 1: Standardize original code
    df["CBSA_CODE_ORIG"] = df["CBSA_CODE"].apply(z5)

    # Step 2: Roll up Metropolitan Divisions to parent MSAs
    df["CBSA_CODE_ROLLED"] = df["CBSA_CODE_ORIG"].map(DIV2CBSA).fillna(df["CBSA_CODE_ORIG"])

    # Step 3: NEW - Apply temporal remappings for codes that changed over time
    # This handles cases like:
    #   - 30100 (Lebanon pre-2013) → 30150 (Lebanon-Claremont current)
    #   - 35154 (New Brunswick MDiv 2013-2023) → 35620 (NY-Newark MSA)
    df["CBSA_CODE_TEMPORAL"] = df["CBSA_CODE_ROLLED"].replace(TEMPORAL_REMAPPINGS)

    # Step 4: Apply legacy renames (e.g., Dayton, other historical changes)
    df["CBSA_CODE_CANON"] = df["CBSA_CODE_TEMPORAL"].replace(LEGACY_RENAMES)

    # Step 5: Flag if original was a Metropolitan Division
    df["is_mdiv"] = df["CBSA_CODE_ORIG"].isin(DIV2CBSA.keys()).astype(int)

    # Step 6: Flag if code was temporally remapped
    df["is_temporal_remap"] = df["CBSA_CODE_ORIG"].isin(TEMPORAL_REMAPPINGS.keys()).astype(int)

    # Step 7: Add comprehensive exclusion flags
    df = add_scope_flags(df, "CBSA_CODE_CANON")

    # Step 8: Store original metro name from BRFSS
    if "MMSANAME" in df.columns:
        df["BRFSS_Metro_Name_Orig"] = df["MMSANAME"]

    # Step 9: Drop original CBSA_CODE to avoid confusion
    df = df.drop(columns=["CBSA_CODE"], errors="ignore")

    return df


# ============================================================================
# SECTION 4: ADD MSA/MDIV NAMES FROM OMB
# NEW: This adds the name columns you requested
# ============================================================================

def add_msa_names_from_omb(brfss_df, omb_2023_path):
    """
    Add MSA names from OMB List 1 (2023) to BRFSS data
    Creates columns: MDiv_Name, MSA_Name_Current

    Args:
        brfss_df: BRFSS DataFrame (after prep_brfss)
        omb_2023_path: Path to list1_2023.xlsx file

    Returns:
        DataFrame with added name columns
    """
    # Load OMB 2023
    omb = pd.read_excel(omb_2023_path, sheet_name="List 1", dtype=str, header=2)
    omb.columns = [str(c).strip() for c in omb.columns]

    # Standardize codes
    if "CBSA Code" in omb.columns:
        omb["CBSA Code"] = omb["CBSA Code"].apply(z5)
    if "Metropolitan Division Code" in omb.columns:
        omb["Metropolitan Division Code"] = omb["Metropolitan Division Code"].apply(z5)

    # Create MDiv name lookup
    mdiv_names = {}
    if "Metropolitan Division Code" in omb.columns and "Metropolitan Division Title" in omb.columns:
        mdiv_data = omb[["Metropolitan Division Code", "Metropolitan Division Title"]].dropna()
        mdiv_data = mdiv_data.drop_duplicates(subset=["Metropolitan Division Code"])
        mdiv_names = dict(mdiv_data.values)

    # Create CBSA name lookup (for parent MSAs)
    cbsa_names = {}
    if "CBSA Code" in omb.columns and "CBSA Title" in omb.columns:
        cbsa_data = omb[["CBSA Code", "CBSA Title"]].dropna()
        cbsa_data = cbsa_data.drop_duplicates(subset=["CBSA Code"])
        cbsa_names = dict(cbsa_data.values)

    # Add MDiv name if original code was a MDiv
    brfss_df["MDiv_Name"] = brfss_df.apply(
        lambda row: mdiv_names.get(row["CBSA_CODE_ORIG"]) if row["is_mdiv"] == 1 else pd.NA,
        axis=1
    )

    # Add current MSA name (from canonical code)
    brfss_df["MSA_Name_Current"] = brfss_df["CBSA_CODE_CANON"].map(cbsa_names)

    return brfss_df


# ============================================================================
# SECTION 5: UPDATED MERGE FUNCTION
# Replace your existing merge_brfss_mefi function with this
# ============================================================================

def merge_brfss_mefi(brfss_df, mefi_df, year_label="2017"):
    """
    Merge BRFSS data with MEFI scores and add comprehensive join status flags

    UPDATED: Now handles exclusions and includes all component scores (1A-3C) and BEA population
    """
    out = brfss_df.merge(mefi_df, left_on="CBSA_CODE_CANON", right_on="CBSA_CODE", how="left")

    # Check for MEFI overall and area scores
    mefi_core = [f"MEFI{year_label}_overall", f"MEFI{year_label}_area1",
                 f"MEFI{year_label}_area2", f"MEFI{year_label}_area3"]
    mefi_core_existing = [c for c in mefi_core if c in out.columns]

    out["mefi_joined"] = out[mefi_core_existing].notna().any(axis=1).astype(int)

    # Updated join_status to handle exclusions
    out["join_status"] = np.where(
        out["mefi_joined"].eq(1),
        "joined",
        np.where(
            out["exclude_from_analysis"].eq(1),
            "excluded",
            "unmatched"
        )
    )

    # Updated mefi_join_note with more detail
    out["mefi_join_note"] = np.where(
        out["mefi_joined"].eq(1),
        "MEFI attached",
        np.where(
            out["scope_note"].ne(""),
            out["scope_note"],
            "No MEFI match — check mapping"
        )
    )

    return out


# ============================================================================
# SECTION 6: MAIN EXECUTION EXAMPLE
# Replace your main execution block with this pattern
# ============================================================================

if __name__ == "__main__":

    print("=" * 80)
    print("APPLYING COMPREHENSIVE BRFSS-MEFI REFINEMENT PATCH")
    print("WITH TEMPORAL REMAPPINGS")
    print("=" * 80)

    # ========= File paths (UPDATE THESE) =========
    SMART2017 = "/content/drive/MyDrive/Saksham_MEFI/SMART_2017_individual.csv"
    SMART2022 = "/content/drive/MyDrive/Saksham_MEFI/SMART_2022_individual.csv"
    MEFI_XLSX = "/content/drive/MyDrive/Saksham_MEFI/MEFI+2022+dataset+for+researchers,+1-5-23 (4).xlsx"
    OMB_2023 = "/content/drive/MyDrive/Saksham_MEFI/list1_2023.xlsx"

    # ========= Load BRFSS data =========
    print("\nStep 1: Loading BRFSS data...")
    B17 = pd.read_csv(SMART2017, dtype=str, low_memory=False)
    B22 = pd.read_csv(SMART2022, dtype=str, low_memory=False)
    print(f"  2017: {len(B17):,} rows")
    print(f"  2022: {len(B22):,} rows")

    # ========= Apply comprehensive prep =========
    print("\nStep 2: Applying comprehensive CBSA standardization...")
    B17 = prep_brfss(B17)
    B22 = prep_brfss(B22)
    print("  ✓ Applied MDiv rollup (42 mappings)")
    print("  ✓ Applied temporal remappings (2 codes: 30100→30150, 35154→35620)")
    print("  ✓ Applied legacy renames")
    print("  ✓ Applied exclusion flags (603 micros, 14 PR/territories, 1 historic)")

    # ========= Add MSA/MDiv names =========
    print("\nStep 3: Adding MSA/MDiv names from OMB...")
    B17 = add_msa_names_from_omb(B17, OMB_2023)
    B22 = add_msa_names_from_omb(B22, OMB_2023)
    print("  ✓ Added MDiv_Name column")
    print("  ✓ Added MSA_Name_Current column")

    # ========= Load and prep MEFI data =========
    print("\nStep 4: Loading MEFI panel data...")
    scores = pd.read_excel(MEFI_XLSX, sheet_name="scores (panel)", header=0)
    scores = scores.rename(columns={
        "CBSA Code":"CBSA_CODE",
        "Area 1: Govern-ment Spending":"Area1_GovtSpending",
        "Area 2: Taxes":"Area2_Taxes",
        "Area 3: Labor Market Regulation":"Area3_Labor",
        "Primary\nState":"Primary_State"
    })
    scores["CBSA_CODE"] = scores["CBSA_CODE"].apply(z5)

    MEFI2017 = scores[scores["year"]==2017].copy()
    MEFI2017["Metro_Area_Name"] = MEFI2017.get("Unnamed: 2", pd.Series([pd.NA]*len(MEFI2017)))
    MEFI2017["Metro_Area_Name_N"] = MEFI2017["Metro_Area_Name"].apply(norm_name)

    MEFI2017_main = (MEFI2017
        .rename(columns={
            "Overall":"MEFI2017_overall",
            "Area1_GovtSpending":"MEFI2017_area1",
            "Area2_Taxes":"MEFI2017_area2",
            "Area3_Labor":"MEFI2017_area3"
        })
        [["CBSA_CODE","Metro_Area_Name","Metro_Area_Name_N",
          "MEFI2017_overall","MEFI2017_area1","MEFI2017_area2","MEFI2017_area3"]]
        .drop_duplicates(subset=["CBSA_CODE"])
    )

    print(f"  Extracted {len(MEFI2017_main)} metro areas from panel data")

    # ========= Extract component scores (1A-3C) =========
    print("\nStep 4b: Extracting component scores (1A-3C)...")
    parts = []
    for sheet, label in [("1A","1A"),("1B","1B"),("1C","1C"),
                         ("2A","2A"),("2B","2B"),("2C","2C"),
                         ("3A","3A"),("3B","3B"),("3C","3C")]:
        print(f"  Extracting component {label}...")
        parts.append(pull_component_by_name(MEFI_XLSX, sheet, label, year=2017))

    MEFI2017_comps = parts[0]
    for t in parts[1:]:
        MEFI2017_comps = MEFI2017_comps.merge(t, on="Metro_Area_Name_N", how="outer")

    print(f"  Combined components: {len(MEFI2017_comps)} unique metro areas")

    # ========= Extract BEA population data =========
    print("\nStep 4c: Extracting BEA population data...")
    try:
        bea = pd.read_excel(MEFI_XLSX, sheet_name="BEA", header=[0,1])

        # Find population column for 2017
        pop_col = None
        for col in bea.columns:
            if isinstance(col, tuple):
                col_combined = ' '.join([str(c).lower() for c in col])
                if "total population" in col_combined or "population" in str(col[0]).lower():
                    if "2017" in str(col[1]) or "2017" in col_combined:
                        pop_col = col
                        print(f"  Found population column: {col}")
                        break

        # Find metro area name column
        name_col = None
        for col in bea.columns:
            if isinstance(col, tuple):
                col_combined = ' '.join([str(c).lower() for c in col])
                for part in col:
                    if "metro area" in str(part).lower() or "metro area" in col_combined:
                        name_col = col
                        print(f"  Found metro name column: {col}")
                        break
                if name_col:
                    break
            else:
                if "metro area" in str(col).lower():
                    name_col = col
                    print(f"  Found metro name column: {col}")
                    break

        if pop_col is not None and name_col is not None:
            BEA2017 = bea[[name_col, pop_col]].copy()
            BEA2017.columns = ["Metro_Area_Name","BEA_pop_2017"]
            BEA2017 = BEA2017[pd.to_numeric(BEA2017["BEA_pop_2017"], errors="coerce").notna()].copy()
            BEA2017["BEA_pop_2017"] = BEA2017["BEA_pop_2017"].astype(float)
            BEA2017["Metro_Area_Name_N"] = BEA2017["Metro_Area_Name"].apply(norm_name)
            MEFI2017_main = MEFI2017_main.merge(BEA2017[["Metro_Area_Name_N","BEA_pop_2017"]],
                                                on="Metro_Area_Name_N", how="left")
            print(f"  Added population data for {BEA2017['BEA_pop_2017'].notna().sum()} metro areas")
        else:
            print("  Warning: Could not find population columns in BEA sheet")
    except Exception as e:
        print(f"  Warning: Could not extract BEA data: {e}")

    # ========= Merge component scores with main MEFI data =========
    print("\nStep 4d: Merging component scores with main MEFI data...")
    MEFI2017_main = MEFI2017_main.merge(MEFI2017_comps, on="Metro_Area_Name_N", how="left")
    print(f"  Final MEFI dataset: {len(MEFI2017_main)} metro areas with all components")

    # ========= Merge BRFSS with MEFI =========
    print("\nStep 5: Merging BRFSS with MEFI...")
    BRFSS2017_MEFI = merge_brfss_mefi(B17, MEFI2017_main, "2017")
    BRFSS2022_MEFI = merge_brfss_mefi(B22, MEFI2017_main, "2017")

    # ========= Report results =========
    print("\n" + "=" * 80)
    print("RESULTS SUMMARY")
    print("=" * 80)

    for year, df in [("2017", BRFSS2017_MEFI), ("2022", BRFSS2022_MEFI)]:
        print(f"\n{year}:")
        print(f"  Total rows: {len(df):,}")
        print(f"\n  Join Status:")
        for status, count in df["join_status"].value_counts().items():
            pct = 100 * count / len(df)
            print(f"    {status:15s}: {count:8,} ({pct:5.1f}%)")

        print(f"\n  Exclusion Breakdown:")
        print(f"    Micropolitan:      {(df['is_micro']==1).sum():,}")
        print(f"    PR/Territory:      {(df['is_pr']==1).sum():,}")
        print(f"    Historic MSA:      {(df['is_excluded_historic']==1).sum():,}")
        print(f"    MDivs rolled:      {(df['is_mdiv']==1).sum():,}")
        print(f"    Temporal remaps:   {(df['is_temporal_remap']==1).sum():,}")

    # ========= Column availability report =========
    print("\n" + "=" * 80)
    print("MEFI COLUMNS AVAILABLE IN OUTPUT FILES")
    print("=" * 80)

    # Check which MEFI columns are available
    mefi_columns = {
        "Overall Score": "MEFI2017_overall",
        "Area 1 (Govt Spending)": "MEFI2017_area1",
        "Area 2 (Taxes)": "MEFI2017_area2",
        "Area 3 (Labor Market)": "MEFI2017_area3",
        "Component 1A": "MEFI2017_1A",
        "Component 1B": "MEFI2017_1B",
        "Component 1C": "MEFI2017_1C",
        "Component 2A": "MEFI2017_2A",
        "Component 2B": "MEFI2017_2B",
        "Component 2C": "MEFI2017_2C",
        "Component 3A": "MEFI2017_3A",
        "Component 3B": "MEFI2017_3B",
        "Component 3C": "MEFI2017_3C",
        "BEA Population (2017)": "BEA_pop_2017",
    }

    print("\nColumns in merged dataset:")
    for label, col in mefi_columns.items():
        if col in BRFSS2017_MEFI.columns:
            non_null = BRFSS2017_MEFI[col].notna().sum()
            print(f"  ✓ {label:30s} ({non_null:,} metro areas)")
        else:
            print(f"  ✗ {label:30s} (not available)")


    # ========= Temporal remapping report =========
    print("\n" + "=" * 80)
    print("TEMPORAL REMAPPING SUMMARY")
    print("=" * 80)
    print("\nThe following codes were remapped due to temporal changes:")
    for old_code, new_code in TEMPORAL_REMAPPINGS.items():
        for year, df in [("2017", BRFSS2017_MEFI), ("2022", BRFSS2022_MEFI)]:
            count = (df["CBSA_CODE_ORIG"] == old_code).sum()
            if count > 0:
                is_excluded = new_code in MICROPOLITAN_AREAS
                status = "excluded (micropolitan)" if is_excluded else "joined to MEFI"
                print(f"  {old_code} → {new_code} ({year}): {count:,} rows — {status}")

    # ========= Save outputs =========
    print("\n" + "=" * 80)
    print("SAVING REFINED DATA")
    print("=" * 80)

    out_dir = "/content/drive/MyDrive/Saksham_MEFI/"

    for year, df in [("2017", BRFSS2017_MEFI), ("2022", BRFSS2022_MEFI)]:
        print(f"\nProcessing {year}:")

        # CLEAN FILE: Only rows that successfully joined to MEFI
        df_clean = df[df["join_status"] == "joined"].copy()
        clean_filename = f"{out_dir}BRFSS{year}_MEFI_CLEAN.csv"
        df_clean.to_csv(clean_filename, index=False)
        print(f"  ✓ Clean file ({len(df_clean):,} rows): BRFSS{year}_MEFI_CLEAN.csv")
        print(f"    - Contains only rows that successfully joined to MEFI")

        # SEMI-CLEAN FILE: Joined + Unmatched (excludes micropolitan/PR/historic)
        df_semiclean = df[df["join_status"].isin(["joined", "unmatched"])].copy()
        semiclean_filename = f"{out_dir}BRFSS{year}_MEFI_SEMICLEAN.csv"
        df_semiclean.to_csv(semiclean_filename, index=False)
        print(f"  ✓ Semi-clean file ({len(df_semiclean):,} rows): BRFSS{year}_MEFI_SEMICLEAN.csv")
        print(f"    - Contains joined ({(df_semiclean['join_status']=='joined').sum():,}) + unmatched ({(df_semiclean['join_status']=='unmatched').sum():,})")
        print(f"    - Excludes: micropolitan, PR/territories, historic MSAs")

        # FULL FILE: All rows for reference
        full_filename = f"{out_dir}BRFSS{year}_MEFI_FULL.csv"
        df.to_csv(full_filename, index=False)
        print(f"  ✓ Full file ({len(df):,} rows): BRFSS{year}_MEFI_FULL.csv")
        print(f"    - Contains all rows with exclusion flags for transparency")

    # ========= Mystery codes report =========
    print("\n" + "=" * 80)
    print("MYSTERY CODES STATUS")
    print("=" * 80)

    if len(MYSTERY_CODES_TODO) == 0:
        print("\n  ✨ All mystery codes have been resolved!")
        print("  • 30100 → remapped to 30150 (micropolitan, excluded)")
        print("  • 35154 → remapped to 35620 (joins to MEFI!)")
        print("  • 39100 → Poughkeepsie (excluded)")
        print("  • 43524 → Washington DC MDiv (joins to MEFI)")
        print("  • 47894 → Washington DC MDiv (joins to MEFI)")
    else:
        print("\n  Remaining mystery codes:")
        for code in MYSTERY_CODES_TODO:
            for year, df in [("2017", BRFSS2017_MEFI), ("2022", BRFSS2022_MEFI)]:
                count = (df["CBSA_CODE_ORIG"] == code).sum()
                if count > 0:
                    sample = df[df["CBSA_CODE_ORIG"] == code].iloc[0]
                    metro_name = sample.get("BRFSS_Metro_Name_Orig", "N/A")
                    print(f"\n  Code {code} ({year}): {count:,} rows")
                    print(f"    Metro name in BRFSS: {metro_name}")
                    print(f"    → TODO: Search OMB for this name manually")

    print("\n" + "=" * 80)
    print("✨ PATCH COMPLETE!")
    print("=" * 80)
    print("\nOutput files created per year (2017 & 2022):")
    print("  1. BRFSS{year}_MEFI_CLEAN.csv")
    print("     → Only rows that successfully joined to MEFI")
    print("     → Use this for main analysis")
    print("\n  2. BRFSS{year}_MEFI_SEMICLEAN.csv")
    print("     → Joined + Unmatched rows")
    print("     → Excludes micropolitan/PR/historic areas")
    print("     → Use for investigating unmatched codes")
    print("\n  3. BRFSS{year}_MEFI_FULL.csv")
    print("     → All rows with exclusion flags")
    print("     → Use for transparency and verification")
    print("\nData improvements:")
    print("  • Component scores (1A-3C) included")
    print("  • BEA population data included")
    print("  • Area scores (1, 2, 3) included")
    print("  • 30100 (2,066 rows) → excluded as micropolitan")
    print("  • 35154 (2,235 rows) → JOINED to MEFI via 35620!")
    print("  • 39100 (517 rows) → excluded (Poughkeepsie)")
    print("  • 43524 & 47894 → Washington DC (joined to MEFI)")
````
